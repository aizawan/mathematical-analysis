## フーリエ級数の近似区間
フーリエ級数展開・複素フーリエ級数展開は，指定した区間上の関数を周期的に拡張し，基本角周波数 $\omega_0 = 2\pi / T$ の整数倍の角周波数を持つ波の重ね合わせで表現する方法であった．

具体例として，$y=t$ のフーリエ級数が区間外でどのような周期的拡張を表すかを，SymPyのフーリエ級数展開を用いて可視化する．

`````python
from sympy import symbols, pi, fourier_series
from sympy.plotting import plot

t = symbols('t')
f = t # f(t) = tの定義

ff = fourier_series(f, (t, -pi, pi)) # フーリエ級数展開を実行
ff_approx = ff.truncate(20) # 第20項目まで計算

plot(f, ff_approx, xlim=(-3*pi, 3*pi), ylim=(-4, 4)) # 区間外まで可視化
`````

このように，区間 $[-\pi,\pi]$ 上の $f(t)=t$ が周期 $2\pi$ で繰り返されている．しかし，与えられた関数や信号が周期性を持つとは限らない．そこで，離散的な角周波数を連続的な角周波数へ拡張し，主に非周期関数を実数全体で解析する **フーリエ変換** を学ぶ．

## フーリエ変換
**フーリエ変換** は，主に非周期関数を連続的な角周波数成分の重ね合わせで表現するフーリエ解析の方法である．時間の関数を角周波数の関数へ変換することで，与えられた関数にどのような角周波数成分が含まれているかを解析できる．一方，角周波数の関数から時間の関数へ戻す **逆フーリエ変換** も定義されている．フーリエ変換は

$$
F \left ( \omega \right ) = \int^{\infty}_{-\infty} f\left ( t \right ) e^{-i\omega t} dt
$$

として定義され，

逆フーリエ変換は

$$
f(t) = \frac{1}{2 \pi} \int^{\infty}_{-\infty} F \left ( \omega \right ) e^{i \omega t} d \omega
$$

で与えられる．

### 区分求積法からのフーリエ変換の導出
では，フーリエ変換と逆フーリエ変換を複素フーリエ級数と複素フーリエ係数から導出する．複素フーリエ級数と複素フーリエ係数は以下で定義されていた．

$$
\begin{align}
f(t) &= \sum_{k=-\infty}^{\infty} C_k e^{ik \omega_0 t} \\
C_k &= \frac{1}{T}\int^{T/2}_{-T/2} f\left ( t \right ) e^{-ik\omega_0 t} dt
\end{align}
$$

ここから積分区間の長さ $T$ を無限に大きくしたときの複素フーリエ級数展開を考える．上記の二式に含まれる基本角周波数 $\omega_0$ について，その整数倍の角周波数 $k \omega_0$ を $\omega_k$ と表記する．

$$
\omega_k = k \omega_0, k = 0, \pm 1, \pm 2, \cdots
$$

$\omega_k$ を使って複素フーリエ級数と係数を整理する．複素フーリエ級数は

$$
\begin{align}
f(t) &= \sum_{k=-\infty}^{\infty} C_k e^{ik \omega_0 t} \\
&= \sum_{k=-\infty}^{\infty} C_k e^{i \omega_k t} \\
\end{align}
$$

となり，複素フーリエ係数は，

$$
\begin{align}
C_k &= \frac{1}{T}\int^{T/2}_{-T/2} f\left ( t \right ) e^{-ik\omega_0 t} dt \\
&= \frac{1}{T}\int^{T/2}_{-T/2} f\left ( t \right ) e^{-i \omega_k t} dt
\end{align}
$$

となる．ここで，$\omega_{k+1}$ と $\omega_k$ は $\omega_0$ の整数倍であることから次のような性質を導ける．

$$
\omega_{k+1}-\omega_k=\omega_0=\frac{2\pi}{T}
$$

この差を $\Delta \omega$ として定義する．

$$
\Delta \omega = \omega_{k+1}-\omega_k=\omega_0=\frac{2\pi}{T}
$$

ここで，有限区間 $[-T/2,T/2]$ 上の積分を

$$
F_T(\omega_k) := \int_{-T/2}^{T/2} f(t) e^{-i \omega_k t} dt
$$

とすると，

$$
C_k = \frac{F_T(\omega_k)}{T} = \frac{F_T(\omega_k)}{2\pi} \Delta \omega
$$

となる．そして，$f(t)$ に $C_k$ を代入する．

$$
\begin{align}
f(t) &= \sum_{k=-\infty}^{\infty} C_k e^{i \omega_k t} \\
&=\sum_{k=-\infty}^{\infty} \frac{F_T\left(\omega_k\right)}{2 \pi} e^{i \omega_k t} \Delta \omega
\end{align}
$$

ここで重要なのは，区分求積法の形式

$$
\int_{-\infty}^{\infty} F(\omega) e^{i \omega t} d\omega \approx \sum_{k=-\infty}^{\infty} F(\omega_k) e^{i \omega_k t} \Delta \omega
$$

に近づいている点である．

そこで，積分区間の長さ $T$ を無限大にする場合を考える．先ほどの式は

$$
f(t) = \frac{1}{2\pi} \sum_{k=-\infty}^{\infty} F_T(\omega_k) e^{i \omega_k t} \Delta \omega
$$

と整理され，これはまさに，

$$
\int_{-\infty}^{\infty} F(\omega) e^{i \omega t} d\omega
$$

のリーマン和に対応する．$T \to \infty$ のとき，積分区間は実数全体へ広がり，$\Delta\omega=2\pi/T \to 0$ となるため，離散的な角周波数の総和が連続的な角周波数の積分へ移ると解釈できる．ここでの議論は，フーリエ級数からフーリエ変換への移行をリーマン和によって説明するものである．


```{admonition} 区分求積法
$$
\int_{a}^{b} f \left ( x \right ) dx = \displaystyle \lim_{N \to \infty} \sum_{k=0}^{N-1} f \left ( x_k \right ) \Delta x
$$

無限区間 $(-\infty,\infty)$ では広義積分として扱う必要があるが，ここではリーマン和と積分の対応を理解するために用いている．
```

積分区間の長さ $T$ を無限大にすると，$F_T(\omega_k)$ は実数全体上の積分へ移る．時刻 $t$ の関数 $f(t)$ を角周波数 $\omega$ の関数 $F(\omega)$ へ変換する操作を **フーリエ変換** と呼び，以下で与えられる．

$$
F \left ( \omega \right ) = \int^{\infty}_{-\infty} f\left ( t \right ) e^{-i\omega t} dt
$$

フーリエ変換は式の通り，$-\infty$ から $\infty$ までのすべての時刻 $t$ にわたる積分で角周波数の関数へ変換できる．
一方で，角周波数 $\omega$ の関数 $F(\omega)$ から時刻 $t$ の関数 $f(t)$ への変換を **逆フーリエ変換** と呼び，以下で与えられる．

$$
f(t) = \frac{1}{2 \pi} \int^{\infty}_{-\infty} F \left ( \omega \right ) e^{i \omega t} d \omega
$$

式の通り，$-\infty$ から $\infty$ までのすべての角周波数 $\omega$ を持つ複素指数関数の重ね合わせと解釈できる．

```{admonition} フーリエ変換の条件と絶対収束
関数 $f$ が絶対可積分，すなわち

$$
f\in L^1(\mathbb R), \qquad \int_{-\infty}^{\infty}|f(t)|dt<\infty
$$

を満たすとする．ここで，$L^1(\mathbb R)$ は，実数全体で絶対値の積分が有限となる関数からなる空間であり，

$$
L^1(\mathbb R)=\left\{f:\mathbb R\to\mathbb C \mathrel{}\middle|\mathrel{} \int_{-\infty}^{\infty}|f(t)|dt<\infty\right\}
$$

と表される．したがって，$f\in L^1(\mathbb R)$ は $f$ が絶対可積分であることを意味する．

このとき，フーリエ変換

$$
F(\omega)=\int_{-\infty}^{\infty}f(t)e^{-i\omega t}dt
$$

は絶対収束する．

ここで，オイラーの公式より複素指数関数は

$$
e^{-i\omega t}=\cos \omega t-i\sin \omega t
$$

と表される．したがって，その絶対値は

$$
\begin{align}
|e^{-i\omega t}|
&=\sqrt{\cos^2\omega t+\sin^2\omega t} \\
&=1
\end{align}
$$

となる．直感的には，$e^{-i\omega t}$ は複素平面の単位円上を回転する長さ $1$ の矢印である．時間 $t$ によって矢印の向きは変化するが，その長さは変化しない．そのため，$f(t)$ に $e^{-i\omega t}$ をかけても大きさは変わらず，

$$
|f(t)e^{-i\omega t}|=|f(t)|\,|e^{-i\omega t}|=|f(t)|
$$

となる．よって，

$$
\int_{-\infty}^{\infty}|f(t)e^{-i\omega t}|dt=\int_{-\infty}^{\infty}|f(t)|dt<\infty
$$

である．したがって，$f\in L^1(\mathbb R)$ なら，すべての $\omega$ に対してフーリエ変換が通常の積分として有限値に定まる．
```

### デルタ関数を用いた形式的な導出
複素フーリエ級数では，離散的な角周波数を持つ複素指数関数を重ね合わせて関数を表現していた．複素フーリエ級数と複素フーリエ係数は以下で定義されていた．

$$
\begin{align}
f(t) &= \sum_{k=-\infty}^{\infty} C_k e^{ik \omega_0 t} \\
C_k &= \frac{1}{T}\int^{T/2}_{-T/2} f\left ( t \right ) e^{-ik\omega_0 t} dt
\end{align}
$$

この複素フーリエ級数の関数系は $\{e^{ik\omega_0 t} \mid k\in\mathbb{Z}\}$ であった．ここで，連続的な角周波数 $\omega$ を持つ関数系 $\{e^{i\omega t} \mid \omega\in\mathbb{R}\}$ を考える．異なる角周波数を持つ複素指数関数を重ねて積分すると，それぞれの振動は互いに打ち消し合い，同じ角周波数の成分だけが残る．この性質はデルタ関数を使って次のように表される．

$$
\begin{align}
\int_{-\infty}^{\infty} e^{i\omega_1 t} \overline{e^{i\omega_2 t}} dt &= \int_{-\infty}^{\infty} e^{i\omega_1 t} e^{-i\omega_2 t} dt \\
&= \int_{-\infty}^{\infty} e^{i (\omega_1-\omega_2) t} dt \\
&= 2 \pi \delta (\omega_1-\omega_2)
\end{align}
$$

左辺の積分は通常の広義積分として収束するものではなく，右辺のデルタ関数を表す形式的な式として解釈する．この関係を，複素指数関数の **一般化された直交性** という．ここで $\delta$ は **ディラックのデルタ関数** とよばれる超関数である．

```{admonition} デルタ関数
デルタ関数は通常の関数ではなく，積分を通して働く超関数である．適切な関数 $f(x)$ に対して

$$
\int_{-\infty}^{\infty} f(x) \delta(x-a) dx = f(a)
$$

を満たす．直感的には，$\delta(x-a)$ は $x=a$ 以外で $0$ となり，全体を積分すると

$$
\int_{-\infty}^{\infty} \delta(x-a) dx = 1
$$

となるものと考えられる．また，複素指数関数との間に

$$
\delta(x) = \frac{1}{2\pi} \int_{-\infty}^{\infty} e^{i\omega x} d\omega
$$

という関係を持つ．この式も通常の積分ではなく，超関数の意味で解釈する．
```

この関係を利用すると，$\omega_1 -\omega_2 \neq 0$ のときは形式的に

$$
\begin{align}
\int_{-\infty}^{\infty} e^{i\omega_1 t} \overline{e^{i\omega_2 t}} dt = 2 \pi \delta (\omega_1-\omega_2) = 0
\end{align}
$$

となり，異なる角周波数の成分は互いに打ち消し合うことがわかる．これにより，離散的な角周波数 $k\omega_0$ を連続値 $\omega$ へ拡張できる．

これにしたがって，複素フーリエ級数

$$
\begin{align}
f(t) &= \sum_{k=-\infty}^{\infty} C_k e^{ik \omega_0 t}
\end{align}
$$

は角周波数が連続値になったことにより，総和を積分に置き換える必要がある．そこで，

$$
f(t) = \int_{-\infty}^{\infty} G(\omega) e^{i\omega t} d\omega
$$

と表し，両辺に $e^{-i\omega t}$ をかける．積分変数と取り出す角周波数を区別するため，積分変数には $s$ を用いる．

$$
f(t)e^{-i\omega t} = \left\{\int_{-\infty}^{\infty} G(s) e^{is t} ds\right\} e^{-i\omega t}
$$

次に，両辺を積分する．

$$
\begin{align}
\int_{-\infty}^{\infty}f(t)e^{-i\omega t}dt &= \int_{-\infty}^{\infty}\left\{\int_{-\infty}^{\infty} G(s) e^{is t} ds\right\} e^{-i\omega t}dt \\
&= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} G(s) e^{is t} e^{-i\omega t} dt ds\\
&= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} G(s) e^{i (s - \omega) t} dt ds\\
&= \int_{-\infty}^{\infty} G(s) \left\{\int_{-\infty}^{\infty}  e^{i (s - \omega) t} dt \right\} ds \\
&= 2\pi \int_{-\infty}^{\infty} G(s) \delta (s - \omega) ds \\
&= 2 \pi G(\omega)
\end{align}
$$

よって，

$$
G(\omega) = \frac{1}{2\pi} \int_{-\infty}^{\infty}f(t)e^{-i\omega t}dt
$$

ここで，

$$
F(\omega) = 2 \pi G(\omega)
$$

とすると，

$$
F(\omega) = \int_{-\infty}^{\infty}f(t)e^{-i\omega t}dt
$$

が得られ，これが関数 $f(t)$ の **フーリエ変換** となる．また，

$$
\begin{align}
f(t) &= \int_{-\infty}^{\infty} G(\omega) e^{i\omega t} d\omega \\
&= \frac{1}{2\pi}\int_{-\infty}^{\infty} F(\omega) e^{i\omega t} d\omega
\end{align}
$$

が得られ，これは **逆フーリエ変換** といわれる．

```{admonition} 積分順序交換に関する注意
二重積分では，常に積分順序を交換できるとは限らない．例えば，被積分関数 $h(s,t)$ が

$$
\iint_{\mathbb R^2}|h(s,t)|dt ds<\infty
$$

を満たす場合は，フビニの定理によって積分順序を交換できる．

今回の形式計算では $h(s,t)=G(s)e^{i(s-\omega)t}$ であり，$|e^{i(s-\omega)t}|=1$ であるため，この絶対可積分条件は一般に満たされない．したがって，通常のフビニの定理による順序交換ではなく，デルタ関数を用いた形式的な計算として解釈する．厳密化する場合は，減衰因子を導入してから極限を取る方法などを用いる．
```

```{admonition} 逆フーリエ変換で元の関数に戻るための条件
$f$ とそのフーリエ変換 $F$ がともに絶対可積分，すなわち

$$
f\in L^1(\mathbb R), \qquad F\in L^1(\mathbb R)
$$

を満たすとき，逆フーリエ変換

$$
\frac{1}{2\pi}\int_{-\infty}^{\infty}F(\omega)e^{i\omega t}d\omega
$$

は $f$ の連続点で $f(t)$ に一致する．このように，逆変換で元の関数に戻るためには，フーリエ変換の存在条件に加えて追加の条件が必要となる．
```

### 波の重ね合わせ
フーリエ級数展開，複素フーリエ級数展開，フーリエ変換はすべて時刻 $t$ に関する関数 $f(t)$ を異なる角周波数の波の重ね合わせで表現する方法である．複素フーリエ級数とフーリエ変換を見比べてみると

**複素フーリエ級数展開**

$$
f(t) = \sum_{k=-\infty}^{\infty} C_k e^{ik \omega_0 t}
, \quad C_k = \frac{1}{T}\int^{T/2}_{-T/2} f\left ( t \right ) e^{-ik\omega_0 t} dt
$$

**フーリエ変換・逆フーリエ変換**

$$
f(t) = \frac{1}{2 \pi} \int^{\infty}_{-\infty} F \left ( \omega \right ) e^{i \omega t} d \omega, \quad F \left ( \omega \right ) = \int^{\infty}_{-\infty} f\left ( t \right ) e^{-i\omega t} dt
$$

となる．複素フーリエ級数では角周波数を基本角周波数の整数倍として離散的に扱い，フーリエ変換では角周波数 $\omega$ を連続的に扱う．そのため，複素フーリエ級数の角周波数成分を **離散スペクトル**，フーリエ変換の角周波数成分を **連続スペクトル** という．

### 具体例
では，以下の関数をフーリエ変換しよう．

$$
f(t) = \begin{cases}
e^{-t} & 0 \leq t \\
0 & t < 0
\end{cases}
$$

解答はクリックで確認できる．

````{dropdown} 解答
フーリエ変換の式 $F(\omega)$ の式に当てはめて計算する．

$$
\begin{align}
F \left ( \omega \right ) &= \int^{\infty}_{-\infty} f\left ( t \right ) e^{-i\omega t} dt \\
&= \int^{0}_{-\infty} 0 \cdot e^{-i\omega t} dt + \int^{\infty}_{0} e^{-t} e^{-i\omega t} dt \\
&= \int^{\infty}_{0} e^{-\left ( 1 + i\omega \right )t} dt \\
&= \lim_{R \to \infty} \left [ \frac{-1}{1+i\omega} e^{ -\left ( 1 + i\omega \right )t} \right ]^{R}_{0} \\
&= \lim_{R \to \infty} \left ( -\frac{e^{- \left ( 1 + i\omega \right )R}}{1+i\omega} + \frac{1}{1+i\omega} \right ) \\
&= \frac{1}{1 + i\omega} \\
&= \frac{1-i\omega}{1 + \omega^2}
\end{align}
$$

ここで，$|e^{-(1+i\omega)R}|=e^{-R}\to0$ $(R\to\infty)$ を利用している．以上より，$f(t)$ のフーリエ変換 $F(\omega)$ を計算することができた．

````

### Pythonでフーリエ変換
では，Pythonでフーリエ変換を実行する．ここでは，絶対可積分なガウス関数を例に用いる．SymPyはフーリエ変換を解析的に計算するため，関数によっては結果を明示的な式として求められない場合がある．

以下でフーリエ変換する関数を定義する．

`````python
from sympy import symbols, exp
from sympy.plotting import plot

t = symbols('t', real=True)
f = exp(-t**2/2)
f
`````

`````python
plot(f)
`````

SymPyでは `fourier_transform(関数, t, nu)` によってフーリエ変換できる．ただし，SymPyは本文の角周波数 $\omega$ ではなく，通常周波数 $\nu$ を用いて

$$
F(\nu)=\int_{-\infty}^{\infty}f(t)e^{-2\pi i\nu t}dt
$$

と定義している．そこで，SymPyで計算した後に $\nu=\omega/(2\pi)$ を代入し，本文の角周波数による表記へ変換する．解析的に求められない場合は，式の代わりに `FourierTransform` クラスが返される．

`````python
from sympy import fourier_transform, pi, simplify
nu = symbols('nu', real=True)
omega = symbols('omega', real=True)

ft_nu = fourier_transform(f, t, nu)
ft = simplify(ft_nu.subs(nu, omega / (2*pi)))
ft
`````

時刻の関数から角周波数の関数に変換されていることを確認するため，フーリエ変換後のグラフをプロットする．

`````python
plot(ft)
`````

SymPyの `inverse_fourier_transform` も通常周波数 $\nu$ の規約を用いる．そこで，$F(\omega)$ に $\omega=2\pi\nu$ を代入してSymPyの規約へ戻してから逆フーリエ変換する．解析的に求められない場合は，式の代わりに `InverseFourierTransform` クラスが返される．

`````python
from sympy import inverse_fourier_transform
ift = inverse_fourier_transform(ft.subs(omega, 2*pi*nu), nu, t)
ift
`````

同様に，角周波数の関数から時刻の関数に戻っていることを確認するため，逆フーリエ変換後のグラフをプロットする．

`````python
plot(ift)
`````

## フーリエ変換の性質
フーリエ変換には微分方程式の適用や角周波数解析において役立つ性質がいくつかある．

### フーリエ変換の線形性
関数 $f(t)$ から角周波数に関する関数 $F(\omega)$ を得るフーリエ変換を $\mathcal{F}(f(t))$ とする．

$$
F \left ( \omega \right ) = \mathcal{F}(f(t)) = \int^{\infty}_{-\infty} f\left ( t \right ) e^{-i\omega t} dt
$$

このときフーリエ変換には以下の線形性の性質がある．

**$f(t)$ の定数倍 $cf(t)$ のフーリエ変換の性質**

$$
\mathcal{F}(cf(t)) = c \mathcal{F}(f(t))
$$

証明はクリックで確認できる．

````{dropdown} 解答
$$
\begin{align}
\mathcal{F}(cf(t)) &= \int^{\infty}_{-\infty} c f\left ( t \right ) e^{-i\omega t} dt \\
&= c \int^{\infty}_{-\infty} f\left ( t \right ) e^{-i\omega t} dt \\
&= c \mathcal{F}(f(t))\\
\end{align}
$$
````

**$f(t)$ と　$g(t)$ の和に関するフーリエ変換の性質**

$$
\mathcal{F}(f(t) \pm g(t)) = \mathcal{F}(f(t)) \pm \mathcal{F}(g(t))
$$

証明はクリックで確認できる．

````{dropdown} 解答
$$
\begin{align}
\mathcal{F}(f(t) \pm g(t)) &= \int^{\infty}_{-\infty} \left \{ f\left ( t \right ) \pm g\left ( t \right ) \right \}  e^{-i\omega t} dt \\
&= \int^{\infty}_{-\infty} f\left ( t \right )  e^{-i\omega t} dt \pm \int^{\infty}_{-\infty} g\left ( t \right ) e^{-i\omega t} dt \\
&= \mathcal{F}(f(t)) \pm \mathcal{F}(g(t))
\end{align}
$$
````

### フーリエ変換の導関数の性質
関数 $f(t)$ のフーリエ変換 $\mathcal{F}(f(t))$ と関数 $f(t)$ の $n$ 次導関数 $f^{(n)}(t)$ には以下の関係がある．

$$
\mathcal{F}(f^{(n)}(t)) = (i \omega)^n \mathcal{F}(f(t))
$$

これは，導関数 $f^{(n)}(t)$ のフーリエ変換は，元の関数 $f(t)$ のフーリエ変換に角周波数領域で $(i\omega)^n$ をかけることに等しいことを表している．

部分積分では，積分区間の端から生じる項 $[f(t)e^{-i\omega t}]_{-\infty}^{\infty}$ が現れる．以下では，$|t|$ が大きくなるにつれて $f(t)$ と必要な導関数が十分速く $0$ に近づき，この端の項が $0$ になると仮定する．

証明はクリックで確認できる．

````{dropdown} 解答
$$
\begin{align}
\mathcal{F}(f'(t)) &= \int^{\infty}_{-\infty} f' \left ( t \right ) e^{-i\omega t} dt\\
&= \left [ f \left ( t \right ) e^{-i \omega t} \right ]^{\infty}_{-\infty} - \int^{\infty}_{-\infty} f \left ( t \right )  \left ( - i \omega \right) e^{-i\omega t} dt\\
&= i \omega \int^{\infty}_{-\infty} f \left ( t \right ) e^{-i\omega t} dt\\
&= i \omega \mathcal{F}(f(t))
\end{align}
$$

第二行目の端の項は，上で仮定した無限遠での減衰によって $0$ となる．同じ計算を $f'(t), f''(t),\ldots$ に繰り返すと，$n$ 次導関数の性質を導出できる．
````

### フーリエ変換の時刻と角周波数のシフト則
**フーリエ変換の時刻のシフト則**

関数 $f(t)$ を $t_0$ だけ平行移動した関数 $f(t-t_0)$ のフーリエ変換は次の性質がある．

$$
\mathcal{F}(f(t-t_0)) = e^{-i \omega t_0} \mathcal{F}(f(t))
$$

証明はクリックで確認できる．

````{dropdown} 解答
$$
\begin{align}
\mathcal{F}(f(t-t_0)) &= \int^{\infty}_{-\infty} f\left ( t - t_0 \right ) e^{-i\omega t} dt \\
&= \int^{\infty}_{-\infty} f\left ( s \right ) e^{-i\omega \left ( s + t_0 \right)} ds \\
&= \int^{\infty}_{-\infty} f\left ( s \right ) e^{-i\omega s} e^{-i\omega t_0 } ds \\
&= e^{-i\omega t_0 } \int^{\infty}_{-\infty} f\left ( s \right ) e^{-i\omega s } ds \\
&= e^{-i\omega t_0 } \int^{\infty}_{-\infty} f\left ( t \right ) e^{-i\omega t } dt \\
&= e^{-i \omega t_0} \mathcal{F}(f(t))
\end{align}
$$
````

**フーリエ変換の角周波数のシフト則**

関数 $f(t)$ のフーリエ変換後の $F(\omega)$ を $\omega_0$ だけ平行移動した関数 $F(\omega-\omega_0)$ には以下の性質がある．

$$
\mathcal{F}(f(t) e^{i \omega_0 t}) = F(\omega - \omega_0)
$$

証明はクリックで確認できる．

````{dropdown} 解答
$$
\begin{align}
\mathcal{F}(f(t) e^{i \omega_0 t}) &= \int^{\infty}_{-\infty} f\left ( t \right ) e^{i \omega_0 t } e^{-i\omega t } dt \\ 
&= \int^{\infty}_{-\infty} f\left ( t \right ) e^{-i \left ( \omega - \omega_0 \right ) t } dt \\ 
&= F(\omega - \omega_0) 
\end{align}
$$
````

### フーリエ変換の時間スケーリング
関数 $f(t)$ の時刻を $a$ 倍した関数 $f(at)$ を考える．$a\neq0$ のとき，フーリエ変換には以下の関係がある．

$$
\mathcal{F}(f(at))=\frac{1}{|a|}F\left(\frac{\omega}{a}\right)
$$

この性質は，時間方向に波形を圧縮すると角周波数領域ではスペクトルが広がり，時間方向に波形を引き伸ばすとスペクトルが狭くなることを表している．

証明はクリックで確認できる．

````{dropdown} 解答
まず $a>0$ として，$s=at$ と変数変換する．このとき $dt=ds/a$ であるから，

$$
\begin{align}
\mathcal{F}(f(at)) &= \int_{-\infty}^{\infty} f(at)e^{-i\omega t}dt \\
&= \frac{1}{a}\int_{-\infty}^{\infty} f(s)e^{-i(\omega/a)s}ds \\
&= \frac{1}{a}F\left(\frac{\omega}{a}\right)
\end{align}
$$

$a<0$ のときは変数変換によって積分区間の向きが反転するため，係数は $1/a$ ではなく $1/|a|$ となる．したがって，$a\neq0$ に対して上記の関係が成り立つ．
````

### 畳み込み定理
二つの関数 $f(t),g(t)$ の **畳み込み** を

$$
(f*g)(t)=\int_{-\infty}^{\infty}f(\tau)g(t-\tau)d\tau
$$

で定義する．$f(t),g(t)$ のフーリエ変換をそれぞれ $F(\omega),G(\omega)$ とすると，畳み込みのフーリエ変換は

$$
\mathcal{F}(f*g)(\omega)=F(\omega)G(\omega)
$$

となる．つまり，時間領域での畳み込みは角周波数領域での積に対応する．以下では，$f,g\in L^1(\mathbb R)$ とし，積分が絶対収束して積分順序を交換できると仮定する．

証明はクリックで確認できる．

````{dropdown} 解答
$$
\begin{align}
\mathcal{F}(f*g)(\omega)
&=\int_{-\infty}^{\infty}\left\{\int_{-\infty}^{\infty}f(\tau)g(t-\tau)d\tau\right\}e^{-i\omega t}dt \\
&=\int_{-\infty}^{\infty}f(\tau)\left\{\int_{-\infty}^{\infty}g(t-\tau)e^{-i\omega t}dt\right\}d\tau
\end{align}
$$

内側の積分で $s=t-\tau$ とおくと，

$$
\begin{align}
\int_{-\infty}^{\infty}g(t-\tau)e^{-i\omega t}dt
&=\int_{-\infty}^{\infty}g(s)e^{-i\omega(s+\tau)}ds \\
&=e^{-i\omega\tau}G(\omega)
\end{align}
$$

である．したがって，

$$
\begin{align}
\mathcal{F}(f*g)(\omega)
&=G(\omega)\int_{-\infty}^{\infty}f(\tau)e^{-i\omega\tau}d\tau \\
&=F(\omega)G(\omega)
\end{align}
$$

となる．
````

一方，時間領域での積は角周波数領域での畳み込みに対応し，

$$
\mathcal{F}(f(t)g(t))(\omega)=\frac{1}{2\pi}(F*G)(\omega)
$$

となる．ここで，

$$
(F*G)(\omega)=\int_{-\infty}^{\infty}F(s)G(\omega-s)ds
$$

である．

証明はクリックで確認できる．

````{dropdown} 解答
$f(t)$ の逆フーリエ変換を代入すると，

$$
\begin{align}
\mathcal{F}(f(t)g(t))(\omega)
&=\int_{-\infty}^{\infty}f(t)g(t)e^{-i\omega t}dt \\
&=\frac{1}{2\pi}\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}F(s)e^{ist}g(t)e^{-i\omega t}dsdt \\
&=\frac{1}{2\pi}\int_{-\infty}^{\infty}F(s)\left\{\int_{-\infty}^{\infty}g(t)e^{-i(\omega-s)t}dt\right\}ds \\
&=\frac{1}{2\pi}\int_{-\infty}^{\infty}F(s)G(\omega-s)ds \\
&=\frac{1}{2\pi}(F*G)(\omega)
\end{align}
$$
````

### Parseval・Plancherelの関係
関数 $f(t),g(t)$ のフーリエ変換をそれぞれ $F(\omega),G(\omega)$ とすると，適切な可積分条件のもとで

$$
\int_{-\infty}^{\infty}f(t)\overline{g(t)}dt
=\frac{1}{2\pi}\int_{-\infty}^{\infty}F(\omega)\overline{G(\omega)}d\omega
$$

が成り立つ．これは，時間領域の内積が角周波数領域の内積に対応することを表している．Plancherelの定理により，この関係は二乗可積分な関数からなる空間 $L^2(\mathbb R)$ にも拡張される．

証明はクリックで確認できる．

````{dropdown} 解答
$g(t)$ の逆フーリエ変換から，

$$
\overline{g(t)}=\frac{1}{2\pi}\int_{-\infty}^{\infty}\overline{G(\omega)}e^{-i\omega t}d\omega
$$

である．これを左辺に代入して積分順序を交換すると，

$$
\begin{align}
\int_{-\infty}^{\infty}f(t)\overline{g(t)}dt
&=\frac{1}{2\pi}\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}f(t)\overline{G(\omega)}e^{-i\omega t}d\omega dt \\
&=\frac{1}{2\pi}\int_{-\infty}^{\infty}\overline{G(\omega)}\left\{\int_{-\infty}^{\infty}f(t)e^{-i\omega t}dt\right\}d\omega \\
&=\frac{1}{2\pi}\int_{-\infty}^{\infty}F(\omega)\overline{G(\omega)}d\omega
\end{align}
$$
````

特に $g=f$ とすると，

$$
\int_{-\infty}^{\infty}|f(t)|^2dt
=\frac{1}{2\pi}\int_{-\infty}^{\infty}|F(\omega)|^2d\omega
$$

となる．この関係は，時間領域と角周波数領域で関数のエネルギーが保存されることを表している．
