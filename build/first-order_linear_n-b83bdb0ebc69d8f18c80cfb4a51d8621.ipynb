このページでは，次の標準形で表される一階線形微分方程式を扱う．

$$
\begin{align}
\frac{dy}{dx}+p(x)y=q(x)
\end{align}
$$

ここで $p(x)$ と $q(x)$ は，考えている区間で定義された $x$ の関数である．特に $q(x)$ が恒等的に $0$ でないとき，この微分方程式を **一階線形非同次微分方程式** と呼ぶ．$q(x)$ がある点で $0$ になることはあってもよいが，$q(x)\equiv 0$ の場合は一階線形同次微分方程式である．

## 特殊解が既知のときの解法
一階線形非同次微分方程式の特殊解が1つわかっているならば，その特殊解を用いて同次微分方程式に帰着できる．与えられた一階線形非同次微分方程式を

$$
\begin{align}
\frac{dy}{dx}+p(x)y=q(x)
\end{align}
$$

とする．その特殊解を $z=z(x)$ とすると，$z$ は次の方程式を満たす．

$$
\begin{align}
\frac{dz}{dx}+p(x)z=q(x)
\end{align}
$$

これらの等式を引き算すると，

$$
\begin{align}
\left(\frac{dy}{dx}+p(x)y\right) - \left(\frac{dz}{dx}+p(x)z\right) &=q(x) - q(x) \\
\left(\frac{dy}{dx} - \frac{dz}{dx}\right) + p(x)\left(y - z\right) &= 0
\end{align}
$$

となる．ここで $Y = y - z$ とおくと，$\frac{dY}{dx} = \frac{dy}{dx} - \frac{dz}{dx}$ より，

$$
\begin{align}
\frac{dY}{dx} + p(x)Y &= 0
\end{align}
$$

が得られる．これは一階線形同次微分方程式であり，任意定数 $C$ を用いて

$$
\begin{align}
Y &= C e^{-\int p(x)\, dx}
\end{align}
$$

と表される．$Y = y - z$ より，本来求めたい関数 $y$ について整理すると，

$$
\begin{align}
y - z &= C e^{-\int p(x)\, dx} \\
y &= C e^{-\int p(x)\, dx} + z
\end{align}
$$

となる．つまり，一階線形非同次微分方程式の一般解は，対応する同次方程式の一般解と，非同次方程式の特殊解の和として表される．

```{admonition} 解の重ね合わせ
一階線形非同次微分方程式の特殊解が1つ与えられている場合，対応する同次方程式の一般解を加えることで，非同次方程式の一般解を表すことができる．

一般に，以下の一階線形微分方程式

$$
\frac{dy}{dx} + p(x)y = q(x)
$$

が与えられたとき，対応する同次方程式

$$
\frac{dy}{dx} + p(x)y = 0
$$

の一般解を $u_h(x)$，非同次方程式の特殊解を $v_p(x)$ とすれば，非同次方程式の一般解 $y(x)$ は次のように表される．

$$
y(x) = u_h(x) + v_p(x)
$$

これは，非同次方程式の任意の2つの解の差が，対応する同次方程式の解になることに対応している．
```

### 特殊解が既知のときの解法の具体例
では，次の一階線形非同次微分方程式を解いてみよう．ただし，特殊解として $z=x-1$ が与えられている．

$$
\begin{align}
\frac{dy}{dx}+y=x
\end{align}
$$

解答はクリックで確認できる．

````{dropdown} 解答
実際，$z=x-1$ とすると $\frac{dz}{dx}=1$ であり，

$$
\begin{align}
\frac{dz}{dx}+z = 1 + (x-1)=x
\end{align}
$$

となるので，$z=x-1$ は与えられた微分方程式の特殊解である．非同次微分方程式の一般解は，同次方程式の一般解と特殊解の和で表される．対応する同次方程式

$$
\begin{align}
\frac{dy}{dx}+y=0
\end{align}
$$

の一般解は，

$$
\begin{align}
y = Ce^{-x}
\end{align}
$$

である．したがって，求める一般解は

$$
\begin{align}
y = Ce^{-x} + x - 1
\end{align}
$$

となる．
````

### 解の重ね合わせの可視化
上の具体例では，対応する同次方程式の解は $y_h=Ce^{-x}$，非同次方程式の特殊解は $y_p=x-1$ であった．このとき

$$
\begin{align}
y = y_h + y_p
\end{align}
$$

が非同次方程式 $\frac{dy}{dx}+y=x$ の解になる．ここで足しているのは，非同次方程式の解同士ではなく，対応する同次方程式の解と非同次方程式の特殊解であることに注意する．まず SymPy で一般解を求め，続いて3つの関数を同じ図に描き，最後に $y_h+y_p$ が元の微分方程式を満たすことを確認する．

`````python
from sympy import symbols, Eq, Derivative, Function, exp, diff, simplify, dsolve

x, C = symbols('x C')
y = Function('y')(x)

# 微分方程式 y' + y = x を SymPy で解く
eq = Eq(Derivative(y, x) + y, x)
general_solution = dsolve(eq, y)

general_solution
`````

ここではグラフが複雑になりすぎないように $C=1$ とし，同次方程式の解（homogeneous solution），特殊解（particular solution），そしてその和（重ね合わせ，superposition）をプロットする．下記のように，同じ図で見ると，特殊解を基準に同次解の分だけずれた曲線が，非同次方程式の別の解になっていることを確認しやすい．

`````python
from sympy import plot

y_h = C * exp(-x)
y_p = x - 1
y_h_c1 = y_h.subs(C, 1)
y_sum_c1 = y_h_c1 + y_p

p = plot(
    y_h_c1,
    y_p,
    y_sum_c1,
    (x, -2, 2),
    title='superposition: homogeneous + particular',
    xlabel='x',
    ylabel='y',
    legend=True,
    show=False,
)
p[0].label = 'homogeneous: exp(-x)'
p[0].line_color = 'blue'
p[1].label = 'particular: x - 1'
p[1].line_color = 'red'
p[2].label = 'sum: exp(-x) + x - 1'
p[2].line_color = 'green'
p.show()
`````

そして，この同次解と特殊解の和が元の非同次方程式を満たすことを下記のように確認できる．

`````python
y_h = C * exp(-x)
y_p = x - 1
y_sum = y_h + y_p
residual = simplify(diff(y_sum, x) + y_sum - x)
residual
`````

出力が $0$ であるので，$y_h+y_p$ を元の微分方程式に代入したときに左辺と右辺が一致することを意味する．

## 特殊解が未知のときの解法
前節では特殊解 $z$ があらかじめ与えられている場合を扱った．しかし，多くの場合は特殊解が事前にはわからないため，微分方程式そのものから特殊解を構成する必要がある．

その代表的な方法が **積分因子法** と **定数変化法** である．積分因子法では，方程式の左辺が積の微分になるような因子を両辺にかける．一方，定数変化法では，対応する同次方程式の一般解に含まれる定数を関数に置き換え，特殊解を求める．どちらの方法も最終的に同じ一般解を与える．

### 積分因子による解法
この形式の微分方程式は **積分因子法** で解くことが一般的である．$P(x)$ を $p(x)$ の1つの原始関数，すなわち $P'(x)=p(x)$ とし，積分因子を

$$
\begin{align}
\mu(x)=e^{P(x)}=e^{\int p(x)\, dx}
\end{align}
$$

とおく．

両辺に積分因子 $\mu(x)$ をかけると，

$$
\begin{align}
\left(\frac{dy}{dx}+p(x)y\right) \mu(x) &= q(x)\mu(x) \\
\frac{dy}{dx}\mu(x) + p(x)y\mu(x) &= q(x)\mu(x)
\end{align}
$$

となる．ここで $\mu'(x)=p(x)\mu(x)$ なので，左辺は積の微分法則より

$$
\begin{align}
\frac{d}{dx}\left(y\mu(x)\right) &= \frac{dy}{dx}\mu(x) + y\mu'(x) \\
&= \frac{dy}{dx}\mu(x) + p(x)y\mu(x)
\end{align}
$$

と書ける．したがって，微分方程式は

$$
\begin{align}
\frac{d}{dx}\left(y\mu(x)\right) &= q(x)\mu(x)
\end{align}
$$

に変形される．両辺を $x$ で積分すると，

$$
\begin{align}
y\mu(x) &= \int q(x)\mu(x)\, dx + C \\
y &= \frac{1}{\mu(x)}\left( \int q(x)\mu(x)\, dx + C \right)
\end{align}
$$

となる．$\mu(x)=e^{\int p(x)\, dx}$ を戻せば，一般解は

$$
\begin{align}
y &= e^{-\int p(x)\, dx} \left( \int q(x)e^{\int p(x)\, dx}\, dx + C \right)
\end{align}
$$

である．ここで $C$ は任意定数である．なお，$\int p(x)\,dx$ に含まれる積分定数は，最終的な任意定数 $C$ に吸収できる．

```{admonition} 積分因子をこのようにおく気持ち
左辺を「ある積の微分」の形にできれば，両辺を積分して解ける．そこで，未知の因子 $\mu(x)$ をかけて，$\frac{d}{dx}(y\mu(x))=\mu(x)\frac{dy}{dx}+\mu'(x)y$ となる形を狙う．元の左辺に $\mu(x)$ をかけた式と見比べると，$\mu'(x)=p(x)\mu(x)$ となればよい．この関係を満たすものが $\mu(x)=e^{\int p(x)\,dx}$ である．
```

#### 積分因子による解法の具体例
では，次の一階線形非同次微分方程式を解いてみよう．以下の微分方程式は $p(x)=1, q(x)=x$ の場合である．

$$
\begin{align}
\frac{dy}{dx}+y=x
\end{align}
$$

解答はクリックで確認できる．

````{dropdown} 解答
まず積分因子を計算する．ここで $p(x)=1$ より，

$$
\begin{align}
\mu(x)=e^{\int p(x)\, dx} = e^{\int 1\, dx} = e^x
\end{align}
$$

となる．両辺に積分因子をかけると，

$$
\begin{align}
\left(\frac{dy}{dx}+y\right) e^x &= xe^x \\
\frac{dy}{dx} e^x + y e^x &= xe^x
\end{align}
$$

であり，左辺は関数 $ye^x$ の微分である．したがって，

$$
\begin{align}
\frac{d}{dx}\left(ye^x\right) &= xe^x
\end{align}
$$

と書ける．両辺を $x$ で積分し，$y$ について整理する．

$$
\begin{align}
ye^x &= \int xe^x\, dx \\
&= xe^x - \int e^x\, dx \\
&= xe^x - e^x + C \\
y &= e^{-x} \left( xe^x - e^x + C \right) \\
&= Ce^{-x} + x - 1
\end{align}
$$

となり，一般解を導出できた．ここで，$C$ は任意定数である．
````

### 定数変化法による解法
一階線形非同次微分方程式は，対応する同次方程式の解に含まれる定数を関数に置き換える **定数変化法** でも解くことができる．

まず，同次方程式

$$
\begin{align}
\frac{dy}{dx}+p(x)y=0
\end{align}
$$

の一般解は

$$
\begin{align}
y_h = C e^{-\int p(x)\, dx}
\end{align}
$$

である．$P(x)=\int p(x)\, dx$ とおき，定数 $C$ を $x$ の関数 $A(x)$ に置き換えた形

$$
\begin{align}
y_p = A(x)e^{-P(x)}
\end{align}
$$

を特殊解として仮定する．

これを非同次方程式に代入すると，

$$
\begin{align}
\frac{d}{dx}\left( A(x)e^{-P(x)} \right)+p(x) A(x)e^{-P(x)} &= q(x) \\
A'(x)e^{-P(x)} + A(x)\left(e^{-P(x)}\right)' + p(x)A(x)e^{-P(x)} &= q(x) \\
A'(x)e^{-P(x)} - A(x)p(x)e^{-P(x)} + p(x)A(x)e^{-P(x)} &= q(x) \\
A'(x)e^{-P(x)} &= q(x) \\
A'(x) &= q(x)e^{P(x)}
\end{align}
$$

となる．右辺は $x$ の関数だけなので，$x$ で積分して

$$
\begin{align}
A(x) &= \int q(x)e^{P(x)}\, dx
\end{align}
$$

を得る．ここでは特殊解を1つ得ればよいので，この不定積分に現れる積分定数は $0$ としてよい．したがって特殊解は

$$
\begin{align}
y_p = e^{-P(x)}\int q(x)e^{P(x)}\, dx
\end{align}
$$

である．最後に，同次方程式の一般解を加えると，非同次方程式の一般解は

$$
\begin{align}
y &= y_h + y_p \\
&= C e^{-P(x)} + e^{-P(x)}\int q(x)e^{P(x)}\, dx \\
&= e^{-\int p(x)\, dx}\left(\int q(x)e^{\int p(x)\, dx}\, dx + C\right)
\end{align}
$$

となる．この解は積分因子による解法と一致している．

```{admonition} 定数変化でこのような解を仮定する気持ち
非同次方程式の解は「同次方程式の解 + 特殊解」と見られる．そこで，同次方程式の解の形 $Ce^{-P(x)}$ を手がかりにし，定数 $C$ だけを関数 $A(x)$ に変えてみる．こうすると，非同次項 $q(x)$ の影響を $A(x)$ の変化として取り込める．実際に代入すると $A'(x)$ だけが残り，特殊解を求められる．
```

#### 定数変化法による解法の具体例
では，次の一階線形非同次微分方程式を解いてみよう．

$$
\begin{align}
\frac{dy}{dx}+y=x
\end{align}
$$

解答はクリックで確認できる．

````{dropdown} 解答
まず，対応する同次方程式

$$
\begin{align}
\frac{dy}{dx}+y=0
\end{align}
$$

の一般解を考える．これは変数分離形で解くことができ，

$$
\begin{align}
y_h = C e^{-x}
\end{align}
$$

となる．定数変化法では，特殊解を

$$
\begin{align}
y_p = A(x)e^{-x}
\end{align}
$$

と仮定する．前述の式変形より，$A'(x) = q(x)e^{\int p(x)\, dx}$ なので，ここでは

$$
\begin{align}
A'(x) &= x e^{\int 1\, dx} \\
&= xe^x
\end{align}
$$

となる．これを $x$ で積分し，特殊解を1つ得るため積分定数を $0$ とすると，

$$
\begin{align}
A(x) &= \int xe^x\, dx \\
&= xe^x - \int e^x\, dx \\
&= xe^x - e^x
\end{align}
$$

である．これを仮定した特殊解 $y_p$ に代入すると，

$$
\begin{align}
y_p &= \left( xe^x - e^x \right)e^{-x} \\
&= x - 1
\end{align}
$$

となる．最後に，非同次方程式の一般解は同次方程式の一般解と特殊解の和であるから，

$$
\begin{align}
y &= y_h + y_p \\
&= Ce^{-x} + x - 1
\end{align}
$$

と一般解を導出できた．ここで，$C$ は任意定数である．
````

## Pythonによる実装
では，具体例に挙げた微分方程式を SymPy で解く．

`````python
from sympy import symbols, Eq, Derivative, Function, dsolve

# 変数と未知関数の定義
x = symbols('x')
y = Function('y')(x)

# dy/dxの定義
dy = Derivative(y, x)

# 微分方程式の定義
eq = Eq(dy + y, x)

# 微分方程式を解く
y_ = dsolve(eq, y)
y_
`````

`````python
import numpy as np
from sympy import plotting
import matplotlib.pyplot as plt

N = 10

eqs = []
for c in np.linspace(-3, 3, N):
    eqs.append(y_.rhs.subs(symbols('C1'), c))
p = plotting.plot(*eqs, (x, -2, 2), show=False)

cm = plt.get_cmap('magma', N)
for i in range(cm.N):
    p[i].line_color = cm(i)
p.show()
`````
