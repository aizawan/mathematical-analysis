## 一階線形微分方程式
ここでは一階線形微分方程式に着目し，その解法を紹介する．一階線形微分方程式は以下の形式で表される微分方程式である．

$$
\begin{align}
\frac{dy}{dx}+p(x)y=q(x)
\end{align}
$$

ここで $p(x)$ と $q(x)$ は，考えている区間で定義された $x$ の関数である．$y$ と $\frac{dy}{dx}$ が1次で現れ，$y^2$ や $y\frac{dy}{dx}$ のような項を含まないことが「線形」の意味である．

特に，$q(x)$ が恒等的に $0$ である場合を同次，恒等的に $0$ でない場合を非同次と呼ぶ．$p(x)$ や $q(x)$ がある点で $0$ になること自体は問題ではない．また，$p(x)\equiv 0$ の場合は直接積分形としても扱える．

```{admonition} 線形微分方程式
線形微分方程式では，未知関数 $y$ に関する微分作用素が線形である．たとえば一階の場合，

$$
L[y]=a_1(x)\frac{dy}{dx}+a_0(x)y
$$

とおくと，任意の定数 $\alpha,\beta$ と関数 $y_1,y_2$ に対して，

$$
\begin{align}
L[\alpha y_1+\beta y_2]=\alpha L[y_1]+\beta L[y_2]
\end{align}
$$

が成り立つ．このような線形作用素を用いて

$$
L[y]=r(x)
$$

と書ける微分方程式を線形微分方程式という．右辺 $r(x)$ が恒等的に $0$ の場合を同次，恒等的に $0$ でない場合を非同次と呼ぶ．このページの標準形では，$L[y]=\frac{dy}{dx}+p(x)y$，$r(x)=q(x)$ である．
```

## 一階線形同次微分方程式
ここでは，$q(x)\equiv 0$ の場合を考える．このとき一階線形微分方程式は以下のように変形できる．

$$
\begin{align}
\frac{dy}{dx}+p(x)y=0 \quad \Rightarrow \quad \frac{dy}{dx}=-p(x)y
\end{align}
$$

この形の微分方程式を **一階線形同次微分方程式** と呼ぶ．

### 一階線形同次微分方程式の解法
上記に示した一階線形同次微分方程式は，変数分離形として解くことができる．まず $y\neq 0$ の区間で，左辺には $y$ に関する項を，右辺には $x$ に関する項がくるように式変形を行う．

$$
\begin{align}
\frac{1}{y} \frac{dy}{dx} &= -p(x)
\end{align}
$$

次に，両辺を $x$ で積分する．

$$
\begin{align}
\int \left(\frac{1}{y} \frac{dy}{dx}\right) dx &= - \int p(x)\, dx \\
\int \frac{1}{y}\, dy &= - \int p(x)\, dx
\end{align}
$$

これを解くと，以下の関係式が得られる．

$$
\begin{align}
\log |y| = - \int p(x)\, dx + C'
\end{align}
$$

ここで，$C'$ は積分定数である．両辺の指数をとると，

$$
\begin{align}
|y| &= e^{-\int p(x)\, dx + C'} \\
y &= \pm e^{C'} e^{-\int p(x)\, dx} \\
&= C e^{-\int p(x)\, dx}
\end{align}
$$

となる．ここで，$C=\pm e^{C'}$ とした．導出の途中では $y$ で割ったため $y=0$ をいったん除外しているが，$C=0$ とすれば $y=0$ も含まれる．したがって，任意定数 $C$ を用いて一般解は

$$
\begin{align}
y(x)=C e^{-\int p(x)\, dx}
\end{align}
$$

と表される．

### 一階線形同次微分方程式の具体例
では，次の一階線形同次微分方程式を解いてみよう．以下の微分方程式は $p(x)=x$ の場合である．

$$
\begin{align}
\frac{dy}{dx}+xy=0
\end{align}
$$

解答はクリックで確認できる．

````{dropdown} 解答
まず，左辺に $y$ のみ，右辺に $x$ の項のみとなるよう変数分離形に整理すると，

$$
\begin{align}
\frac{dy}{dx} &= -xy \\
\frac{1}{y} \frac{dy}{dx} &= -x
\end{align}
$$

となる．続いて，両辺を $x$ で積分する．

$$
\begin{align}
\int \left(\frac{1}{y} \frac{dy}{dx}\right) dx &= - \int x\, dx \\
\int \frac{1}{y}\, dy &= - \int x\, dx
\end{align}
$$

積分定数を $C'$ とすると以下となる．

$$
\begin{align}
\log |y| = - \frac{x^2}{2} + C'
\end{align}
$$

$y$ について整理すると，

$$
\begin{align}
y &= \pm e^{- \frac{x^2}{2} + C'} \\
&= \pm e^{C'} e^{- \frac{x^2}{2}} \\
&= C e^{- \frac{x^2}{2}}
\end{align}
$$

となり，一般解が得られた．ここで，$C$ は任意定数である．
````

### Pythonによる実装
では，上記の微分方程式を SymPy で解く．単一のセルで作成したが，各行での動作は前回の講義のように Step-by-step で確認されたい．

`````python
from sympy import symbols, Eq, Derivative, Function, dsolve

# 変数と未知関数の定義
x = symbols('x')
y = Function('y')(x)

# dy/dxの定義
dy = Derivative(y, x)

# 微分方程式の定義
eq = Eq(dy + x*y, 0)

# 微分方程式を解く
y_ = dsolve(eq, y)
y_
`````

`````python
import numpy as np
from sympy import plotting
import matplotlib.pyplot as plt

N = 10

eqs = []
for c in np.linspace(-3, 3, N):
    eqs.append(y_.rhs.subs(symbols('C1'), c))
p = plotting.plot(*eqs, (x, -2, 2), show=False)

cm = plt.get_cmap('magma', N)
for i in range(cm.N):
    p[i].line_color = cm(i)
p.show()
`````
