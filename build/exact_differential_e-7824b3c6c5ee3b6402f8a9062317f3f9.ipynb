**全微分** とは，多変数関数の各変数の微小変化による関数値の変化を表すものである．2変数関数 $F(x,y)$ の全微分 $dF$ は，偏微分を用いて

$$
dF = \frac{\partial F}{\partial x} dx + \frac{\partial F}{\partial y} dy
$$

と書ける．ここで，次の形の一階微分方程式を考える．

$$
P(x,y)dx+Q(x,y)dy = 0
$$

もしある関数 $F(x,y)$ が存在して，

$$
\frac{\partial F}{\partial x} = P(x,y), \quad \frac{\partial F}{\partial y} = Q(x,y)
$$

を満たすならば，左辺は $dF$ と書ける．このとき

$$
P(x,y)dx + Q(x,y)dy = dF = 0
$$

となるため，この微分方程式を **完全微分方程式** という．ここでの $F(x,y)$ の解は $F(x,y)=C$ という陰関数表示で表される．

## 完全微分方程式の解法
完全微分方程式を解くとは，

$$
\frac{\partial F}{\partial x}=P(x,y), \quad \frac{\partial F}{\partial y}=Q(x,y)
$$

を満たす関数 $F(x,y)$ を求めることである．このような $F$ が見つかれば，

$$
dF = P(x,y)dx + Q(x,y)dy = 0
$$

であるから，解は

$$
F(x,y)=C
$$

となる．ただし，$C$ は任意定数である．

まず，与えられた微分方程式が完全形であることを確認する．ここでは，$P,Q$ の偏導関数が連続で，解を考える範囲に穴や切れ目がないものとする．このような前提では，完全微分方程式であるための判定条件は

$$
\frac{\partial P(x,y)}{\partial y} = \frac{\partial Q(x,y)}{\partial x}
$$

である．この条件を満たすとき，$P(x,y)$ を $x$ について積分して

$$
F(x,y) = \int P(x,y)\, dx + \phi(y)
$$

とおく．$x$ で積分するとき，$y$ は定数のように扱われるため，積分定数に相当する部分は $y$ の関数 $\phi(y)$ として残す．

次に，この式を $y$ について偏微分し，$\frac{\partial F}{\partial y}=Q(x,y)$ と比較する．

$$
\frac{\partial F}{\partial y}
= \frac{\partial}{\partial y} \left\{ \int P(x,y)\, dx \right\} + \frac{d\phi(y)}{dy}
= Q(x,y)
$$

これにより $\frac{d\phi(y)}{dy}$ が求まり，さらに $y$ で積分すれば $\phi(y)$ が決まる．最後に $F(x,y)=C$ とおけば，一般解が得られる．同様に，先に $Q(x,y)$ を $y$ について積分し，$x$ の関数を補う方法でもよい．

```{admonition} 全微分
2変数関数 $F(x,y)$ の全微分 $dF$ は，$x$ 方向の変化と $y$ 方向の変化を合わせたものである．

$$
dF=\frac{\partial F}{\partial x}dx+\frac{\partial F}{\partial y}dy
$$

したがって，$dF=0$ は$F$ の値が変わらないことを表し，解が $F(x,y)=C$ という等高線になると理解できる．
```

```{admonition} 完全形
$P(x,y)dx+Q(x,y)dy$ が，ある関数 $F(x,y)$ の全微分 $dF$ と同じ形で書けるとき，この微分方程式は完全形であるという．

$$
P(x,y)dx+Q(x,y)dy=dF
$$

つまり，$P$ は $F$ を $x$ で偏微分したもの，$Q$ は $F$ を $y$ で偏微分したものになっている．この対応があるかどうかを調べる手がかりが，$\frac{\partial P}{\partial y}=\frac{\partial Q}{\partial x}$ である．
```


## 完全微分方程式の具体例
次の微分方程式を解いてみよう．

$$
\left ( 2x + 3y \right )dx + \left ( 3x + 8y \right )dy = 0
$$

解答はクリックで確認できる．

````{dropdown} 解答
まず，与えられた微分方程式が完全微分方程式であるかを確かめる．$P(x,y)=2x+3y$，$Q(x,y)=3x+8y$ とすると，

$$
\begin{aligned}
\frac{\partial P(x,y)}{\partial y} = 3, \quad
\frac{\partial Q(x,y)}{\partial x} = 3
\end{aligned}
$$

である．したがって

$$
\frac{\partial P(x,y)}{\partial y} =  \frac{\partial Q(x,y)}{\partial x}
$$

が成り立ち，完全微分方程式である．

次に $P(x,y)$ を $x$ について積分する．

$$
\begin{aligned}
F(x,y) &= \int P(x,y)\, dx + \phi(y) \\
&= \int \left( 2x + 3y \right)\, dx + \phi(y) \\
&= x^2 + 3xy + \phi(y)
\end{aligned}
$$

続いて，上記の $F(x,y)$ を $y$ について偏微分する．

$$
\begin{aligned}
\frac{\partial F(x,y)}{\partial y}
&= \frac{\partial}{\partial y} \left( x^2 + 3xy + \phi(y) \right ) \\
&= 3x + \frac{d\phi(y)}{dy}
\end{aligned}
$$

これは $Q(x,y)=3x+8y$ と一致する必要があるので，

$$
\frac{d\phi(y)}{dy}=8y
$$

である．これを $y$ について積分すると，

$$
\begin{aligned}
\phi(y) &= \int 8y\,dy \\
&= 4y^2
\end{aligned}
$$

となる．したがって，

$$
F(x,y)=x^2+3xy+4y^2
$$

であり，任意定数 $C$ を用いて一般解は

$$
\begin{aligned}
x^2 + 3xy + 4y^2 = C
\end{aligned}
$$

と表される．
````


## Pythonによる実装
では，SymPy を使って完全微分方程式を解く．この形式は `dsolve` に直接渡すのではなく，上記の解法のステップを順番に計算していく．まず，ライブラリのインポートと変数，$P(x,y),Q(x,y)$ を定義する．

`````python
from sympy import symbols, Eq, integrate, diff

# 変数の定義
x = symbols('x')
y = symbols('y')

# P(x,y), Q(x,y) の定義
P = 2*x + 3*y
Q = 3*x + 8*y
`````

完全微分方程式であるための条件を確認する．これまでは `Derivative` 関数を使って導関数を定義していたが，今回は `diff` 関数で偏導関数を計算する．`diff` は第一引数に微分する式を，第二引数に微分する変数を与える．

`````python
diff(P, y)
`````

`````python
diff(Q, x)
`````

どちらも $3$ となり，完全形であることがわかった．続いて，$P(x,y)$ を $x$ で積分してみる．積分は `integrate` 関数で実行できる．

`````python
F_x = integrate(P, x)
F_x
`````

$\phi(y)$ については，

$$
\begin{aligned}
\frac{d\phi(y)}{dy} = Q(x,y) - \frac{\partial}{\partial y}\left(\int P(x,y)\,dx\right)
\end{aligned}
$$

であるので，以下を計算する．


`````python
d_phi = Q - diff(F_x, y)
d_phi
`````

これを $y$ について積分する．

`````python
phi = integrate(d_phi, y)
phi
`````

したがって，未知関数 $F(x,y)$ は以下となる．

`````python
F = F_x + phi
F
`````

となり，一般解を導出できた．最後に，$C=4$ として解曲線 $F(x,y)=4$ をプロットする．完全微分方程式の解は $z=F(x,y)$ という曲面そのものではなく，$F(x,y)=C$ を満たす等高線として表される．

`````python
from sympy.plotting import plot_implicit

plot_implicit(Eq(F, 4), (x, -5, 5), (y, -5, 5), aspect_ratio=(1, 1))
`````
