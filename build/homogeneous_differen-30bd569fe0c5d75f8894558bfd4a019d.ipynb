このページで扱う **同次微分方程式** は，一階微分方程式の右辺が $\frac{y}{x}$ の関数として書ける形を指す．つまり，以下のような形式の微分方程式である．

$$
\begin{align}
\frac{dy}{dx} = f\left(\frac{y}{x}\right)
\end{align}
$$

ただし，$x\neq 0$ の区間で $u=\frac{y}{x}$ とおく．$f(u)-u\neq 0$ の範囲では，一般解は以下で与えられる．

$$
\begin{align}
\int \frac{1}{f(u) -u}\,du= \log |x| + C
\end{align}
$$

ただし，$C$は積分定数とする．また，$f(a)=a$ を満たす定数 $a$ がある場合は，$u=a$，すなわち $y=ax$ も解になる．

```{admonition} 同次式
同次式は **斉次式（せいじしき）** ともいう．定数 $\lambda$ について，$0$ 以上の整数 $n$ ，および関数 $f(x,y)$ に対して，

$$
f(\lambda x, \lambda y) = \lambda^n f(x, y)
$$

が成立するとき，関数 $f$ は $n$ 次の同次式という．例えば，

$$
f(x,y)=x^2+xy+y^2
$$

については，

$$
\begin{align}
f(\lambda x, \lambda y)
&=(\lambda x)^2+(\lambda x)(\lambda y)+(\lambda y)^2 \\
&=\lambda^2(x^2+xy+y^2) \\
&=\lambda^2 f(x,y)
\end{align}
$$

となるため，$2$ 次の同次式である．また，$g(x,y)=3x-2y$ は $1$ 次の同次式である．

なお，このページで扱う同次微分方程式は，右辺が $\frac{y}{x}$ の関数として書ける一階微分方程式を指している．したがって，右辺が $0$ でなくてもこの意味では同次形と呼ぶ．一方，線形微分方程式でいう同次・非同次は，非同次項が $0$ かどうかによる別の分類である．
```

## 一般解の導出
同次微分方程式を解くためには，$\frac{y}{x}=u$として変数変換をすれば良い．この変数変換は

$$
\begin{align}
\frac{y}{x}=u \Rightarrow y = xu
\end{align}
$$

であり，これを$x$について微分すると，

$$
\begin{align}
\frac{dy}{dx} = u + x\frac{du}{dx}
\end{align}
$$

のようになる．ここで $u=y/x$ という $x$ の関数なので右辺は積の微分を利用する．そして，$u$について整理すると，

$$
\begin{align}
\frac{du}{dx} = \frac{1}{x} \left(\frac{dy}{dx} -u\right)
\end{align}
$$

となり，$\frac{dy}{dx}=f\left(\frac{y}{x}\right)=f(u)$であるので，次の形になる．

$$
\begin{align}
\frac{du}{dx} = \frac{1}{x} \left(f(u) -u\right)
\end{align}
$$

ここで $x\neq 0$ かつ $f(u)-u\neq 0$ の範囲では，上記の式は $u$ と $x$ のみが含まれ，$u$ と $x$ について以下のように整理すると変数分離形となる．

$$
\begin{align}
\frac{1}{f(u) -u}\,du= \frac{1}{x}\,dx
\end{align}
$$

変数分離形の解法と同様に両辺を積分し，

$$
\begin{align}
\int \frac{1}{f(u) -u}\,du= \int \frac{1}{x}\,dx
\end{align}
$$

以下の一般解が得られる．

$$
\begin{align}
\int \frac{1}{f(u) -u}\,du= \log |x| + C
\end{align}
$$

ただし，$C$は積分定数とする．なお，$f(a)=a$ を満たす定数 $a$ がある場合，変数分離で割った $f(u)-u$ が $0$ となるため，直線解 $y=ax$ を別途確認する．

```{admonition} 積の微分法則
積の微分法則とは $y=f(x)g(x)$ という関数の変数$x$ の微分は以下となる法則である．

$$
\frac{dy}{dx} = f'(x)g(x) + f(x)g'(x)
$$

$u$ の変数変換では，$u$ は $x$ の関数であるので，$y'=u'x + ux'$ という微分になっていることに注意されたい．
```

## 同次微分方程式の具体例
では，次の同次微分方程式を解いてみよう．

$$
\frac{dy}{dx} = \frac{x^2 + y^2}{xy}
$$

解答はクリックで確認できる．

````{dropdown} 解答
一見すると変数分離形とも同次形とも見て取れないが，ここで，右辺の分子と分母を $x^2$ で割ると以下のように $y/x$ の形が現れる．

$$
\frac{dy}{dx} = \frac{1 + \left(\frac{y}{x}\right)^2}{\left(\frac{y}{x}\right)}
$$

ここで $u=\frac{y}{x}$ として変数変換をする．前述より，

$$
\begin{align}
u + x\frac{du}{dx} &= \frac{1 + u^2}{u}
\end{align}
$$

である．これを与えられた微分方程式に代入すると

$$
\begin{align}
\frac{dy}{dx} &= u + x\frac{du}{dx} \\
u + x\frac{du}{dx} &= \frac{1 + u^2}{u} \\
x\frac{du}{dx} &= \frac{1 + u^2}{u} - u \\
x\frac{du}{dx} &= \frac{1}{u} \\
u\,du &= \frac{1}{x}\,dx
\end{align}
$$

と変数分離形となった．変数分離形の解法にしたがって両辺を積分する．

$$
\begin{align}
\int u du &= \int \frac{1}{x} dx \\
\frac{u^2}{2} &= \log |x| + D \\
u^2 &= 2 \log |x| + 2D \\
\end{align}
$$

最後に，$u=\frac{y}{x}$ を代入し，$y$ について整理すると，

$$
\begin{align}
\frac{y^2}{x^2} &= 2 \log |x| + 2D \\
y^2 &= 2 x^2 \log |x| + 2D x^2 \\
y &= \pm x \sqrt{2 \log |x| + C}
\end{align}
$$

として，一般解が得られる．ここで，$D$は積分定数であり，$C=2D$とした．この例では $f(u)-u=\frac{1}{u}$ であり，$u=0$ は元の微分方程式でも定義されないため，直線解の取りこぼしはない．
````

## Pythonによる実装
では，これをsympyで解く．

`````python
from sympy import symbols, Eq, Derivative, Function, dsolve

# 変数，関数，導関数を定義．
x = symbols('x')
y = Function('y')(x)
dy = Derivative(y, x)

# 与えられた同次微分方程式を定義
eq = Eq(dy, (x**2 + y**2) / (x*y))

# 微分方程式を解く．
y_ = dsolve(eq, y)
`````

このとき手計算した一般解からもわかるように，解は正負の2つの枝を持つことに注意する．また，sympy の結果は $\log(x)$ を含むため，ここでは $x>0$ の区間で表示する．具体的には，積分定数 `C1` を $0 \leq C_1 \leq 3$，横軸を $1 \leq x \leq 2$ としている．この範囲では，平方根の中身 $C_1+2\log(x)$ が負にならない．
