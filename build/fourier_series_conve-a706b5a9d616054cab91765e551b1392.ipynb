フーリエ級数展開では，関数 $f(x)$ を三角関数の和で表した．しかし，フーリエ級数の式

$$
\begin{align}
f(x) \sim \frac{a_0}{2}+\sum_{k=1}^{\infty}\left(a_k\cos kx+b_k\sin kx\right)
\end{align}
$$

に現れる記号 $\sim$ は，必ずすべての点で右辺が $f(x)$ と一致することを意味しているわけではない．特に不連続点をもつ関数では，フーリエ級数がどの値へ収束するのかを確認する必要がある．

このノートでは，フーリエ級数の収束先を与える **ディリクレ・ジョルダンの定理** を紹介し，不連続点の近くで現れる **ギブス現象** を Python で確認する．

## フーリエ級数の部分和
無限和を一度に計算することはできないので，実際には有限個の項で打ち切った部分和を考える．区間 $[-\pi,\pi]$ 上の関数 $f(x)$ に対して，$N$ 次の部分和を

$$
\begin{align}
S_N(x)=\frac{a_0}{2}+\sum_{k=1}^{N}\left(a_k\cos kx+b_k\sin kx\right)
\end{align}
$$

と書く．フーリエ級数の収束を考えるとは，$N$ を大きくしたときに $S_N(x)$ がどの値へ近づくかを考えることである．

連続な点では，よい条件のもとで $S_N(x)$ は $f(x)$ に近づく．一方，不連続な点では，$S_N(x)$ はその点で定義された関数値そのものではなく，左右から近づいた値の平均に近づく．

## ディリクレ・ジョルダンの定理
フーリエ級数の収束先を知るための代表的な定理として，ディリクレ・ジョルダンの定理がある．

```{admonition} ディリクレ・ジョルダンの定理
関数 $f(x)$ が周期 $2\pi$ の周期関数であり，区間 $[-\pi,\pi]$ 上で有界変動であるとする．このとき，$f(x)$ のフーリエ級数の部分和 $S_N(x)$ は，各点 $x$ で

$$
\lim_{N\to\infty}S_N(x)=\frac{f(x+0)+f(x-0)}{2}
$$

へ収束する．ここで，$f(x+0)$ は右極限，$f(x-0)$ は左極限を表す．
```

特に，$f(x)$ が点 $x$ で連続であれば $f(x+0)=f(x-0)=f(x)$ なので，フーリエ級数は $f(x)$ に収束する．一方，不連続点では左右極限の平均に収束する．この性質は，フーリエ級数で段差を近似するときに非常に重要である．

有界変動という条件は少し抽象的だが，有限個の点を除いて滑らかで，有限個の不連続点しか持たないような段差のある関数を考えている．

### 具体例：ステップ関数
フーリエ級数展開のノートで扱った次の関数を考える．

$$
f(x)=\begin{cases}
1 & -\pi < x \leq 0 \\
0 & 0 < x \leq \pi
\end{cases}
$$

この関数を周期 $2\pi$ で周期的に拡張して考える．前回求めたフーリエ級数は

$$
\begin{align}
f(x) \sim \frac{1}{2}-\frac{2}{\pi}\sum_{n=1}^{\infty}\frac{1}{2n-1}\sin(2n-1)x
\end{align}
$$

であった．この関数の $x=0$ における左右極限を確認する．$0$ の左側，すなわち $x<0$ で $x$ が十分 $0$ に近いときは $f(x)=1$ である．したがって，

$$
f(0-)=\lim_{x\to -0}f(x)=1
$$

である．一方，$0$ の右側，すなわち $x>0$ で $x$ が十分 $0$ に近いときは $f(x)=0$ である．したがって，

$$
f(0+)=\lim_{x\to +0}f(x)=0
$$

である．よって，ディリクレ・ジョルダンの定理より

$$
\lim_{N\to\infty}S_N(0)=\frac{1+0}{2}=\frac{1}{2}
$$

となる．これは，$f(0)=1$ と定義していたとしても，フーリエ級数の収束先は $1/2$ であることを意味している．

## Pythonによる部分和の可視化
では，部分和 $S_N(x)$ を SymPy で計算して，項数を増やしたときの近似の様子を確認する．まず，ステップ関数を Piecewise で定義し，fourier_series でフーリエ級数を求める．その後，truncate() で有限個の項に打ち切って可視化する．

````python
from sympy import symbols, pi, Piecewise, fourier_series
from sympy.plotting import plot

x = symbols("x")
f = Piecewise((1, x <= 0), (0, True))

ff = fourier_series(f, (x, -pi, pi))
ff
````

````python
ff_1 = ff.truncate(1)
ff_3 = ff.truncate(3)
ff_10 = ff.truncate(10)
ff_50 = ff.truncate(50)

plot(f, ff_1, ff_3, ff_10, ff_50, (x, -pi, pi), ylim=(-0.3, 1.3))
````

項数を増やすと，不連続点から離れた場所では近似がよくなる．一方，不連続点の近くでは振動が残り，段差の上側や下側にはみ出す部分が現れる．この現象を **ギブス現象** という．ギブス現象は「項数が少ないために起こる単なる誤差」ではなく，不連続をフーリエ級数で表すときに本質的に現れる現象である．
