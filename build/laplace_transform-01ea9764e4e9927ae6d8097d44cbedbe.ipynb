## フーリエ変換からラプラス変換へ
フーリエ変換は，主に非周期関数を実数全体で解析する方法であった．関数 $f(t)$ が絶対可積分，すなわち

$$
f\in L^1(\mathbb R), \qquad \int_{-\infty}^{\infty}|f(t)|dt<\infty
$$

を満たすとき，フーリエ変換

$$
F(\omega)=\int_{-\infty}^{\infty}f(t)e^{-i\omega t}dt
$$

は絶対収束する．ただし，絶対可積分性はフーリエ変換が存在するための十分条件であり，必要条件ではない．

例えば，定数関数 $f(t)=1$ は絶対可積分ではない．通常の広義積分としてフーリエ変換を考えると，$\omega\neq0$ に対して

$$
\begin{aligned}
\int_{-R}^{R}e^{-i\omega t}dt
&=\left[\frac{e^{-i\omega t}}{-i\omega}\right]_{-R}^{R} \\
&=\frac{e^{-i\omega R}-e^{i\omega R}}{-i\omega} \\
&=\frac{e^{i\omega R}-e^{-i\omega R}}{i\omega} \\
&=\frac{2i\sin(\omega R)}{i\omega} \\
&=\frac{2\sin(\omega R)}{\omega}
\end{aligned}
$$

となる．$\omega\neq0$ のとき，$R\to\infty$ としても $\sin(\omega R)$ は $-1$ から $1$ の間を振動し続けるため，この式は一つの値へ収束しない．また，$\omega=0$ のときは $\int_{-R}^{R}1\,dt=2R\to\infty$ となる．そこで，時間を $t\geq0$ に限定し，指数関数による重みを掛けることで，定数関数や指数的に増加する関数も扱える **ラプラス変換** を学ぶ．

## ラプラス変換
本講義では，$t\geq0$ で定義された関数を扱う **片側ラプラス変換** を考える．複素変数 $s$ を

$$
s=\sigma+i\omega, \qquad \operatorname{Re}(s)=\sigma
$$

と表す．実数 $\sigma$ を固定し，次の関数を考える．

$$
g_\sigma(t)=\begin{cases}f(t)e^{-\sigma t} & t\geq0 \\ 0 & t<0\end{cases}
$$

$\sigma$ を適切に選んで $g_\sigma$ が絶対可積分になるとき，そのフーリエ変換は

$$
\begin{align}
G_\sigma(\omega)&=\int_{-\infty}^{\infty}g_\sigma(t)e^{-i\omega t}dt \\
&=\int_0^\infty f(t)e^{-(\sigma+i\omega)t}dt
\end{align}
$$

となる．ここで $s=\sigma+i\omega$ とおくことで，関数 $f(t)$ のラプラス変換を

$$
F(s)=\mathcal L\{f(t)\}=\int_0^\infty f(t)e^{-st}dt
$$

と定義する．$f(t)$ を **原関数**，$F(s)$ を **像関数** と呼ぶことがある．

### $\operatorname{Re}(s)$ と収束
複素指数関数は

$$
e^{-st}=e^{-\sigma t}e^{-i\omega t}
$$

と分解できる．オイラーの公式から $e^{-i\omega t}=\cos(\omega t)-i\sin(\omega t)$ なので，

$$
\begin{aligned}
|e^{-st}|
&=|e^{-\sigma t}e^{-i\omega t}| \\
&=e^{-\sigma t}|e^{-i\omega t}| \\
&=e^{-\sigma t}\sqrt{\cos^2(\omega t)+\sin^2(\omega t)} \\
&=e^{-\sigma t}
\end{aligned}
$$

となる．$e^{-i\omega t}$ は複素平面の単位円上で向きを変えるだけで，大きさを変えない．したがって，虚部 $\omega$ は振動を表し，実部 $\sigma=\operatorname{Re}(s)$ が減衰の強さと積分の収束を決める．

```{admonition} ラプラス変換が存在するための十分条件
$f(t)$ が有限区間ごとに区分的に連続で，十分大きな $t$ に対して，ある定数 $M>0$ と $\alpha\in\mathbb R$ を用いて

$$
|f(t)|\leq Me^{\alpha t}
$$

と評価できるとする．このとき，

$$
\begin{aligned}
|f(t)e^{-st}|
&=|f(t)|\,|e^{-st}| \\
&=|f(t)|e^{-\sigma t} \\
&\leq Me^{\alpha t}e^{-\sigma t} \\
&=Me^{-(\sigma-\alpha)t} \\
&=Me^{-(\operatorname{Re}(s)-\alpha)t}
\end{aligned}
$$

となる．$\beta=\operatorname{Re}(s)-\alpha>0$ とおくと，$e^{-\beta t}$ は $e^{-x}\to0$ $(x\to\infty)$ と同じ理由で急速に減衰し，十分大きな $T$ に対して

$$
\int_T^\infty Me^{-\beta t}dt=\frac{M}{\beta}e^{-\beta T}<\infty
$$

となる．比較によってラプラス変換は絶対収束する．単に被積分関数が $0$ に近づくだけでは積分の収束は保証されず，ここでは指数関数が十分速く減衰することが重要である．このような $s$ の範囲を **収束領域** という．
```

```{admonition} 区分的に連続な関数
関数が有限区間 $[0,T]$ 上で **区分的に連続** であるとは，区間を有限個に分けると，各区間の内部で関数が連続となり，分割点における左右の極限が有限値として存在することをいう．有限個の段差を持つステップ関数は区分的に連続である．一方，区間内で値が無限大へ発散する場合などは，この条件を満たさないことがある．

片側ラプラス変換では，この条件をすべての有限区間 $[0,T]$ で満たすことを仮定する．
```

```{admonition} 片側ラプラス変換と積分範囲
片側ラプラス変換は $t\geq0$ の値だけを利用し，$t<0$ の値は積分に含めない．これは，時刻 $t=0$ から始まるシステムや初期値問題を扱う場合に適している．必要であれば，$t<0$ で $0$ となる因果的な関数として解釈できる．

一方，

$$
F(s)=\int_{-\infty}^{\infty}f(t)e^{-st}dt
$$

を **両側ラプラス変換** というが，本講義では片側ラプラス変換を扱う．
```

## 逆ラプラス変換
像関数 $F(s)$ から原関数 $f(t)$ を求める操作を **逆ラプラス変換** といい，$\mathcal L^{-1}\{F(s)\}$ と表す．逆ラプラス変換は，次のBromwichの反転公式で与えられる．

$$
f(t)=\mathcal L^{-1}\{F(s)\}=\frac{1}{2\pi i}\int_{c-i\infty}^{c+i\infty}F(s)e^{st}ds, \quad t>0
$$

右辺の複素積分を **Bromwich積分（ブロムウィッチ積分）** という．積分路は複素平面上の垂直線 $\operatorname{Re}(s)=c$ であり，$c$ はラプラス変換の収束領域内に選ぶ．

微分方程式を解く際は，Bromwich積分を直接計算するのではなく，変換表と部分分数分解を用いて逆ラプラス変換を求めることが多い．

```{admonition} 逆変換を詳しく学ぶための数学
Bromwich積分は複素平面上の積分路に沿う複素積分である．この積分を直接計算し，極や留数との関係まで理解するには，正則関数，複素積分，留数定理などを扱う **複素解析** が必要となる．

一方，逆フーリエ変換は実変数 $\omega$ に関する複素数値関数の積分なので，定義そのものに複素積分は必要ない．逆変換が成立する条件を厳密に理解するには実解析，ルベーグ積分，$L^p$ 空間などのフーリエ解析が必要であり，積分を複素平面へ拡張して計算する場合には複素解析も用いる．
```

````{dropdown} フーリエ逆変換からの導出（補足）
$c$ を収束領域内の実数とし，

$$
g_c(t)=\begin{cases}f(t)e^{-ct} & t\geq0 \\ 0 & t<0\end{cases}
$$

とおく．$g_c(t)$ のフーリエ変換は

$$
G_c(\omega)=\int_0^\infty f(t)e^{-(c+i\omega)t}dt=F(c+i\omega)
$$

である．フーリエ逆変換を適用すると，$t>0$ に対して

$$
f(t)e^{-ct}=\frac{1}{2\pi}\int_{-\infty}^{\infty}F(c+i\omega)e^{i\omega t}d\omega
$$

となる．両辺に $e^{ct}$ を掛けると，

$$
f(t)=\frac{1}{2\pi}\int_{-\infty}^{\infty}F(c+i\omega)e^{(c+i\omega)t}d\omega
$$

を得る．ここで $s=c+i\omega$ と変数変換すると $ds=i\,d\omega$ であるため，

$$
f(t)=\frac{1}{2\pi i}\int_{c-i\infty}^{c+i\infty}F(s)e^{st}ds
$$

が得られる．
````

## ラプラス変換の具体例
主要な関数のラプラス変換と収束条件を確認する．

### ステップ関数
次の単位ステップ関数を考える．片側ラプラス変換では $t\geq0$ の値だけを積分するため，積分区間上では $f(t)=1$ である．

$$
f(t)=\begin{cases}1 & t\geq0 \\ 0 & t<0\end{cases}
$$

````{dropdown} 解答
$$
\begin{align}
F(s)&=\lim_{R\to\infty}\int_0^R e^{-st}dt \\
&=\lim_{R\to\infty}\left[-\frac{e^{-st}}{s}\right]_0^R \\
&=\lim_{R\to\infty}\left(-\frac{e^{-sR}}{s}+\frac{1}{s}\right) \\
&=\frac{1}{s}, \qquad \operatorname{Re}(s)>0
\end{align}
$$

$\operatorname{Re}(s)>0$ なら $|e^{-sR}|=e^{-\operatorname{Re}(s)R}\to0$ となることを利用している．
````

### デルタ関数（補足）
デルタ関数は通常の関数ではなく，積分を通して働く **超関数** である．超関数は各点での値によって定めるのではなく，通常の関数に積分を通して作用した結果によって定める．デルタ関数の場合は，$a>0$ に対して

$$
\int_{-\infty}^{\infty}g(t)\delta(t-a)dt=g(a)
$$

という性質から，

$$
\mathcal L\{\delta(t-a)\}=\int_0^\infty\delta(t-a)e^{-st}dt=e^{-as}
$$

となる．片側ラプラス変換では，$t=0$ のインパルス全体を積分範囲に含める工学的な規約を用いて $\mathcal L\{\delta(t)\}=1$ とすることが多い．デルタ関数と超関数については，[フーリエ変換のノート](../fourier_analysis/fourier_transform.ipynb)でも補足している．

### 指数関数
$a\in\mathbb R$ とし，指数関数 $f(t)=e^{at}$ のラプラス変換を求める．

````{dropdown} 解答
$$
\begin{align}
F(s)&=\lim_{R\to\infty}\int_0^R e^{-(s-a)t}dt \\
&=\lim_{R\to\infty}\left[-\frac{e^{-(s-a)t}}{s-a}\right]_0^R \\
&=\frac{1}{s-a}, \qquad \operatorname{Re}(s)>a
\end{align}
$$
````

### 冪関数
$n=0,1,2,\ldots$ とし，冪関数 $f(t)=t^n$ のラプラス変換を求める．

````{dropdown} 解答
$I_n(s)=\int_0^\infty t^ne^{-st}dt$ とおく．$n\geq1$ かつ $\operatorname{Re}(s)>0$ のとき $\lim_{R\to\infty}R^ne^{-sR}=0$ であるため，部分積分により

$$
\begin{align}
I_n(s)&=\left[-\frac{t^ne^{-st}}{s}\right]_0^\infty+\frac{n}{s}\int_0^\infty t^{n-1}e^{-st}dt \\
&=\frac{n}{s}I_{n-1}(s)
\end{align}
$$

を得る．$I_0(s)=1/s$ からこの関係を繰り返すと，

$$
\mathcal L\{t^n\}=\frac{n!}{s^{n+1}}, \qquad \operatorname{Re}(s)>0
$$

となる．具体的に $n=1,2$ を代入して確認すると，

$$
\begin{aligned}
\mathcal L\{t\}
&=\lim_{R\to\infty}\left(\left[-\frac{te^{-st}}{s}\right]_0^R+\frac{1}{s}\int_0^R e^{-st}dt\right) \\
&=0+\frac{1}{s}\mathcal L\{1\} \\
&=\frac{1}{s^2}
\end{aligned}
$$

および

$$
\begin{aligned}
\mathcal L\{t^2\}
&=\lim_{R\to\infty}\left(\left[-\frac{t^2e^{-st}}{s}\right]_0^R+\frac{2}{s}\int_0^R te^{-st}dt\right) \\
&=0+\frac{2}{s}\mathcal L\{t\} \\
&=\frac{2}{s^3}
\end{aligned}
$$

となる．
````

### 三角関数
$\omega\in\mathbb R$ とし，角周波数 $\omega$ を持つ余弦関数 $f(t)=\cos(\omega t)$ のラプラス変換を求める．

````{dropdown} 解答
$$
\begin{align}
F(s)&=\frac{1}{2}\int_0^\infty e^{-(s-i\omega)t}dt+\frac{1}{2}\int_0^\infty e^{-(s+i\omega)t}dt \\
&=\frac{1}{2}\left(\frac{1}{s-i\omega}+\frac{1}{s+i\omega}\right) \\
&=\frac{s}{s^2+\omega^2}, \qquad \operatorname{Re}(s)>0
\end{align}
$$
````

続いて，正弦関数 $f(t)=\sin(\omega t)$ のラプラス変換を求める．

````{dropdown} 解答
$$
\begin{align}
F(s)&=\frac{1}{2i}\int_0^\infty e^{-(s-i\omega)t}dt-\frac{1}{2i}\int_0^\infty e^{-(s+i\omega)t}dt \\
&=\frac{1}{2i}\left(\frac{1}{s-i\omega}-\frac{1}{s+i\omega}\right) \\
&=\frac{\omega}{s^2+\omega^2}, \qquad \operatorname{Re}(s)>0
\end{align}
$$
````

## ラプラス変換の性質
ラプラス変換には，微分方程式を代数的に解くために役立つ性質がある．以下では $\mathcal L\{f(t)\}=F(s)$ とする．

### ラプラス変換の一意性
$f$ と $g$ がラプラス変換の存在条件を満たし，共通の収束領域で

$$
\mathcal L\{f(t)\}=\mathcal L\{g(t)\}
$$

となるなら，$f(t)=g(t)$ がほとんど至る所で成り立つ．特に $f$ と $g$ が区分的に連続なら，両者はすべての連続点で一致する．孤立した点での値の違いは積分に影響しないため，ラプラス変換からは区別できない．

```{admonition} 一意性の直感と役割
$F(s)$ は一つの $s$ に対する積分値だけではなく，収束領域内のさまざまな $s$ に対する重み付き積分の集まりである．これは元の関数の特徴を記録した識別情報のように考えられる．もし $F(s)=G(s)$ が収束領域全体で成り立つなら，差 $h(t)=f(t)-g(t)$ はすべての重み $e^{-st}$ に対して積分すると $0$ になり，通常の条件のもとでは差が残らない．

この性質により，微分方程式を $s$ 領域の代数方程式へ変換して解いても，得られた $Y(s)$ から元の時間関数 $y(t)$ を一意に復元できる．したがって，変換表や部分分数分解を使った逆ラプラス変換が解法として成立する．
```

### ラプラス変換の線形性
定数 $c_1,c_2$ に対して，

$$
\mathcal L\{c_1f(t)+c_2g(t)\}=c_1\mathcal L\{f(t)\}+c_2\mathcal L\{g(t)\}
$$

が成り立つ．

````{dropdown} 証明
$$
\begin{align}
\mathcal L\{c_1f(t)+c_2g(t)\}
&=\int_0^\infty\{c_1f(t)+c_2g(t)\}e^{-st}dt \\
&=c_1\int_0^\infty f(t)e^{-st}dt+c_2\int_0^\infty g(t)e^{-st}dt \\
&=c_1\mathcal L\{f(t)\}+c_2\mathcal L\{g(t)\}
\end{align}
$$
````

### $s$ 領域の移動則
関数 $f(t)$ のラプラス変換が $F(s)$ のとき，

$$
\mathcal L\{e^{at}f(t)\}=F(s-a)
$$

が成り立つ．

````{dropdown} 証明
$$
\begin{align}
\mathcal L\{e^{at}f(t)\}
&=\int_0^\infty e^{at}f(t)e^{-st}dt \\
&=\int_0^\infty f(t)e^{-(s-a)t}dt \\
&=F(s-a)
\end{align}
$$
````

### 時間スケーリング
関数 $f(t)$ のラプラス変換が $F(s)$ のとき，$a>0$ に対して

$$
\mathcal L\{f(at)\}=\frac{1}{a}F\left(\frac{s}{a}\right)
$$

が成り立つ．

````{dropdown} 証明
$u=at$ と変数変換すると，

$$
\begin{align}
\mathcal L\{f(at)\}
&=\int_0^\infty f(at)e^{-st}dt \\
&=\frac{1}{a}\int_0^\infty f(u)e^{-(s/a)u}du \\
&=\frac{1}{a}F\left(\frac{s}{a}\right)
\end{align}
$$
````

### 導関数のラプラス変換
関数 $f(t)$ の $n$ 次導関数について，

$$
\mathcal L\{f^{(n)}(t)\}
=s^nF(s)-\sum_{k=0}^{n-1}s^{n-1-k}f^{(k)}(0+)
$$

が成り立つ．ここで $f^{(k)}(0+)$ は $t=0$ における右極限である．特に，

$$
\mathcal L\{f'(t)\}=sF(s)-f(0+)
$$

$$
\mathcal L\{f''(t)\}=s^2F(s)-sf(0+)-f'(0+)
$$

となる．

````{dropdown} 一次導関数の場合の証明
先に示した指数位数の十分条件のもとで $\operatorname{Re}(s)>\alpha$ とすると，$\lim_{R\to\infty}f(R)e^{-sR}=0$ となる．部分積分により，

$$
\begin{align}
\mathcal L\{f'(t)\}
&=\lim_{R\to\infty}\int_0^R f'(t)e^{-st}dt \\
&=\lim_{R\to\infty}\left([f(t)e^{-st}]_0^R+s\int_0^R f(t)e^{-st}dt\right) \\
&=sF(s)-f(0+)
\end{align}
$$
````

導関数のラプラス変換では，微分が $s$ に関する代数演算へ置き換わり，初期値も同時に式へ組み込まれる．次のノートでは，この性質を利用して初期値問題を解く．

## Pythonでラプラス変換
SymPyを用いてラプラス変換を実行する．ここでは $f(t)=t^2$ を例に，変換結果と収束条件を確認する．まず，ラプラス変換する関数を定義する．

`````python
from sympy import symbols
from sympy.plotting import plot

t = symbols('t', real=True)
f = t**2
f
`````

`````python
plot(f)
`````

SymPyでは `laplace_transform(関数, t, s)` によってラプラス変換できる．変数 $s$ は複素変数なので，`real=True` は指定しない．

`noconds=True` を指定しない場合，`(変換結果, 収束境界, 追加条件)` の組が返される．収束境界が $a$ のとき，変換は $\operatorname{Re}(s)>a$ で収束する．

`````python
from sympy import laplace_transform

s = symbols('s')
lt, convergence_plane, cond = laplace_transform(f, t, s)
lt, convergence_plane, cond
`````

$f(t)=t^2$ の場合は `(2/s**3, 0, True)` が得られる．したがって，変換結果は $2/s^3$，収束条件は $\operatorname{Re}(s)>0$ であり，追加条件はない．

SymPyは `inverse_laplace_transform(関数, s, t)` による逆ラプラス変換もサポートしている．解析的に求められない場合は，式の代わりに `InverseLaplaceTransform` クラスが返される．

`````python
from sympy import inverse_laplace_transform
ilt = inverse_laplace_transform(lt, s, t)
ilt
`````

`````python
plot(ilt)
`````

逆ラプラス変換の結果は $t^2\theta(t)$ となる．片側ラプラス変換は $t\geq0$ の値だけを利用するため，SymPyは元の関数を $t<0$ で $0$ とした **因果的な延長** を返す．ここで $\theta(t)$ はヘビサイド関数である．

SymPyでは次の規約を用いる．

$$
\theta(t)=\begin{cases}
1 & t>0 \\
\frac{1}{2} & t=0 \\
0 & t<0
\end{cases}
$$

$t=0$ の値は定義の流儀によって異なるが，一点だけの値は通常のラプラス積分には影響しない．

`````python
from sympy import Heaviside
Heaviside(t)
`````

`````python
plot(Heaviside(t))
`````
