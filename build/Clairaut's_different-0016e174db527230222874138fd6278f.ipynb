**クレロー型微分方程式（Clairaut's equation）** は，次の形式で表される一階微分方程式である．

$$
y=x\frac{dy}{dx} + f \left( \frac{dy}{dx}\right)
$$

ここで，関数 $f$ は導関数 $\frac{dy}{dx}$ を変数に持つ関数である．未知関数 $y$ だけでなく導関数も非線形に現れることがあるが，導関数を新しい関数として扱うことで解を調べられる．

## クレロー型微分方程式の解法
クレロー型微分方程式では，導関数

$$
p=\frac{dy}{dx}
$$

を導入する．このとき方程式は

$$
y=px+f(p)
$$

と書ける．ここで $p=p(x)$ と考え，両辺を $x$ で微分する．$y$ と $p$ がどちらも $x$ の関数であることに注意すると，

$$
\begin{aligned}
\frac{dy}{dx}&=p+x\frac{dp}{dx} + f'(p) \frac{dp}{dx} \\
p&=p+ \frac{dp}{dx} \left ( x + f'(p)\right ) \\
0 &= \frac{dp}{dx} \left ( x + f'(p)\right )
\end{aligned}
$$

となる．したがって，次の2つの場合に分かれる．

1つ目は

$$
\frac{dp}{dx}=0
$$

の場合である．このとき $p=C$ は定数なので，元の式に戻すと

$$
y=Cx+f(C)
$$

が得られる．これは任意定数 $C$ を含む直線族であり，クレロー型微分方程式の一般解である．

2つ目は

$$
x+f'(p)=0
$$

の場合である．この条件と元の式

$$
y=px+f(p)
$$

を連立し，パラメータ $p$ を消去できれば，任意定数を含まない解が得られる．この解は一般解の包絡線になり，特異解と呼ばれることがある．ただし，$f$ の形によっては $p$ を消去した形で明示できない場合もある．

## クレロー型微分方程式の具体例
次の微分方程式を解いてみよう．

$$
y=x\frac{dy}{dx} + \left( \frac{dy}{dx}\right)^2
$$

解答はクリックで確認できる．

````{dropdown} 解答
この微分方程式はクレロー型である．$p=\frac{dy}{dx}$ とおくと，

$$
y=px + p^2
$$

となる．ここで $p=p(x)$ と考え，両辺を $x$ で微分する．

$$
\begin{aligned}
\frac{dy}{dx}&=p+x\frac{dp}{dx} + 2p \frac{dp}{dx} \\
p&=p+ \frac{dp}{dx} \left ( x + 2p\right ) \\
0 &= \frac{dp}{dx} \left ( x + 2p\right )
\end{aligned}
$$

この等式が成立するためには，

$$
\frac{dp}{dx}=0
$$

または

$$
x+2p=0
$$

であればよい．まず $\frac{dp}{dx}=0$ の場合，$p=C$ なので，

$$
y=Cx+C^2
$$

という一般解が得られる．これは傾き $C$ を持つ直線族である．

次に $x+2p=0$ の場合，$p=-\frac{x}{2}$ である．これを $y=px+p^2$ に代入すると，

$$
\begin{aligned}
y &= \left ( -\frac{x}{2} \right ) x + \left ( -\frac{x}{2} \right )^2 \\
&= - \frac{1}{4} x^2
\end{aligned}
$$

という特異解が得られる．

```{admonition} 特異解
**特異解** とは，一般解の任意定数 $C$ にどのような値を代入しても得られない解である．特殊解は一般解の任意定数を初期条件で固定した解であり，特異解とは区別する．
```

この特異解は，一般解として得られた直線族すべてに接する曲線であり，**包絡線** と呼ばれる．

```{admonition} 包絡線
**包絡線（envelope）** とは，パラメータを含む曲線族の各曲線に接する曲線である．クレロー型微分方程式では，特異解が一般解の包絡線として現れることが多い．
```
````

## Pythonによる実装
では，上記の微分方程式を SymPy で解く．SymPy は一般解を返すが，クレロー型の特異解は自動では返さない場合があるため，手計算で得た特異解もあわせて確認する．

`````python
from sympy import symbols, Eq, Derivative, Function, dsolve

# 変数と未知関数の定義
x = symbols('x')
y = Function('y')(x)

# dy/dxの定義
dy = Derivative(y, x)

# 微分方程式の定義
eq = Eq(y, dy*x + dy**2)

# 微分方程式を解く
y_ = dsolve(eq, y)
y_
`````

`````python
import numpy as np
from sympy import plotting
import matplotlib.pyplot as plt

N = 10
C1 = symbols('C1')

eqs = []
for c in np.linspace(-1, 1, N):
    eqs.append(y_.rhs.subs(C1, c))
eqs.append(-x**2 / 4)

p = plotting.plot(*eqs, (x, -5, 5), ylim=(-8, 8), show=False)

cm = plt.get_cmap('magma', N)
for i in range(N):
    p[i].line_color = cm(i)
p[N].line_color = 'black'
p[N].label = 'singular solution'
p.show()
`````
