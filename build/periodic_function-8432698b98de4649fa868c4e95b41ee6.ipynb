$\sin x$ のグラフは **正弦波** と呼ばれる代表的な波であり，$\cos x$ も $\sin x$ を位相方向にずらした同じ形の周期的な波である．このように同じ波形が一定間隔で繰り返される関数を **周期関数** と呼ぶ．

関数 $f$ に対して，ある正の定数 $T>0$ が存在し，すべての $x$ について

$$
f(x+T) = f(x)
$$

が成り立つとき，$f$ を周期関数と呼び，このような $T$ を **周期** と呼ぶ．$\sin x$ と $\cos x$ は周期 $T=2\pi$ の周期関数であり，三角関数の性質から，

$$
\sin(x+2\pi) = \sin(x), \quad \cos(x+2\pi) = \cos(x)
$$

が成り立つ．周期のうち最小の正の値が存在するとき，それを **基本周期** と呼ぶ．たとえば $\sin x$ と $\cos x$ の基本周期は $2\pi$ である．一方で，$2T, 3T, ...$ のような正の整数倍も周期であるが，基本周期とは区別する．なお，定数関数のように任意の正の数が周期となり，基本周期を持たない周期関数もある．

また，$\sin kx$ や $\cos kx$ の基本周期は $2\pi/k$ である．$k$ が大きくなるほど，同じ区間に含まれる波の数が増える．この考え方は，後で扱う周波数・角周波数や，フーリエ級数で関数を正弦波の重ね合わせとして表す考え方につながる．

## Pythonによる実装
では，$\sin x$ と $\cos x$ について，周期 $2\pi$ を足しても関数の値が変わらないことを SymPy で確認し，関数をプロットしてみよう．

```python
from sympy import symbols, sin, cos, pi, simplify
from sympy.plotting import plot

x = symbols('x')

sin_periodicity = simplify(sin(x + 2*pi) - sin(x))
cos_periodicity = simplify(cos(x + 2*pi) - cos(x))

print(f"sin(x + 2*pi) - sin(x) = {sin_periodicity}")
print(f"cos(x + 2*pi) - cos(x) = {cos_periodicity}")

plot(
    sin(x), cos(x),
    (x, -3*pi, 3*pi),
);
```

## 周期・周波数・角周波数
周期 $T$ は，波形が 1 回繰り返されるまでの長さを表す．一方，**周波数** は単位長さあたりに波形が何回繰り返されるかを表す量である．周波数を $\nu$ とすると，周期と周波数には

$$
\nu=\frac{1}{T}
$$

という関係がある．周期が短いほど，同じ区間に多くの波が入るため，周波数は大きくなる．

$\sin x$ は長さ $2\pi$ の区間に 1 個の波を持つ．一方，$\sin kx$ は長さ $2\pi$ の区間に $k$ 個の波を持つ．したがって，$k$ は波の細かさを決める係数であり，角度 $x$ に対する **角周波数** とみなすことができる．

時間 $t$ の関数として波を考える場合，正弦波は

$$
\sin(\omega t)
$$

のように書かれる．ここで，$\omega$ を **角周波数** と呼ぶ．この波の周期 $T$ は

$$
\omega T=2\pi
$$

を満たすので，

$$
T=\frac{2\pi}{\omega}, \qquad \nu=\frac{1}{T}=\frac{\omega}{2\pi}
$$

となる．つまり，角周波数と周波数は

$$
\omega=2\pi\nu
$$

で結びついている．フーリエ級数では，基本周期 $T$ に対応する基本角周波数 $\omega_0=2\pi/T$ を考え，$\omega_0, 2\omega_0, 3\omega_0, \ldots$ の角周波数を持つ三角関数を足し合わせて関数を表す．

```python
plot(
    sin(x),
    sin(2*x),
    sin(3*x),
    (x, -pi, pi),
    legend=True,
);
```

## 振幅と位相
正弦波は一般に

$$
f(x)=A\sin(kx+\phi)
$$

のように書くことができる．ここで，$A$ は **振幅**，$\phi$ は **位相** を表す．振幅は波の大きさを決める値であり，$A>0$ のとき $A\sin(kx+\phi)$ は最大値 $A$，最小値 $-A$ をとる．一般には振幅は $|A|$ である．

$k$ は波の繰り返しの細かさを決める値であり，$x$ に対する角周波数に対応する．$k>0$ のとき，$\sin(kx)$ の基本周期は $2\pi/k$ である．一方，$\phi$ は波の横方向のずれを表す．$\sin(kx+\phi)$ は，$\sin(kx)$ を $x$ 軸方向に $-\phi/k$ だけ平行移動した波である．したがって，$\phi>0$ のときグラフは左にずれ，$\phi<0$ のときグラフは右にずれる．

例えば，$2\sin x$ は $\sin x$ より振幅が大きい波であり，$\sin(x+\pi/4)$ は $\sin x$ を左に $\pi/4$ だけずらした波である．

```python
f_base = sin(x)
f_amplitude = 2 * sin(x)
f_phase = sin(x + pi/4)

plot(
    f_base,
    f_amplitude,
    f_phase,
    (x, -2*pi, 2*pi),
    legend=True,
);
```

## 三角関数の合成
三角関数を足し合わせることで，より複雑な周期関数を作ることができる．例えば，

$$
g(x)=\sin x+\frac{1}{2}\sin 2x+\frac{1}{3}\cos 3x
$$

を考える．$\sin x$，$\sin 2x$，$\cos 3x$ はいずれも $2\pi$ を周期として持つため，これらを足し合わせた $g(x)$ も周期 $2\pi$ の周期関数である．ここでは基本周期が $2\pi$ なので，基本角周波数は $\omega_0=1$ である．したがって，$\sin x$ は $\omega_0$，$\sin 2x$ は $2\omega_0$，$\cos 3x$ は $3\omega_0$ の角周波数を持つ成分と見なせる．このように，基本角周波数の整数倍の三角関数を足し合わせることで，基本周期を保ちながら複雑な波形を表現できる．

```python
g = sin(x) + sin(2*x) / 2 + cos(3*x) / 3

plot(g, (x, -3*pi, 3*pi));
```

## 周期的拡張
周期関数でない関数であっても，ある区間で定義された形を繰り返すことで周期関数として扱うことができる．この操作を **周期的拡張** と呼ぶ．

例えば，区間 $[-\pi, \pi)$ 上で定義された関数

$$
f(x)=x \qquad (-\pi \leq x < \pi)
$$

を考える．この関数自体は全実数上では周期関数ではないが，区間 $[-\pi, \pi)$ の形を長さ $2\pi$ ごとに繰り返すことで，周期 $2\pi$ の周期関数を作ることができる．周期的拡張した関数を $f_{\mathrm{per}}(x)$ と書くと，

$$
f_{\mathrm{per}}(x+2\pi)=f_{\mathrm{per}}(x)
$$

が成り立つ．

```python
import numpy as np
import matplotlib.pyplot as plt

# 周期的拡張を表示するため，複数周期にわたる x の値を用意する
x_plot = np.linspace(-4*np.pi, 4*np.pi, 1000)
# 元の関数 f(x)=x を定義する 1 周期分の区間 [-pi, pi) を用意する
# endpoint=False により，次の周期の -pi と重なる pi を含めない
x_base = np.linspace(-np.pi, np.pi, 200, endpoint=False)

# 区間 [-pi, pi) 上では f(x)=x なので，x_base をそのまま関数値とする
f_base = x_base
# x_plot を [-pi, pi) から [0, 2*pi) にずらし，周期 2*pi で割ったあまりをとる．
# すると f_periodic の値は [0, 2*pi) の範囲で同じ形を繰り返す
# そして，最後に -pi を引くことで [-pi, pi) の範囲に戻す
f_periodic = (x_plot + np.pi) % (2*np.pi) - np.pi

plt.figure(figsize=(8, 3))
plt.plot(x_plot, f_periodic, label="periodic extension")
plt.plot(x_base, f_base, linewidth=3, label="f(x)=x on [-pi, pi)")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(-np.pi, color="gray", linestyle="--", linewidth=0.8)
plt.axvline(np.pi, color="gray", linestyle="--", linewidth=0.8)
plt.xlim(-4*np.pi, 4*np.pi)
plt.ylim(-np.pi, np.pi)
plt.legend()
plt.show()
```
