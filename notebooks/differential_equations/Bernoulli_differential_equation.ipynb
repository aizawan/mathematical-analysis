**ベルヌーイ型微分方程式** は，次の形式で表される一階微分方程式である．

$$
\frac{dy}{dx} + p(x)y = q(x)y^n
$$

ここで，$p(x)$ と $q(x)$ は考えている区間で定義された $x$ の関数であり，$n$ は定数である．$n\neq 0,1$ の場合は $y^n$ の項があるため一階非線形微分方程式であるが，適切な変数変換によって一階線形微分方程式に帰着できる．

## ベルヌーイ型微分方程式の解法
ベルヌーイ型微分方程式は，一階非線形微分方程式の中でも特別な形を持つ．$n$ の値によって既に学んだ線形微分方程式になる場合と，変数変換が必要な場合に分かれる．

### $n=0$ の場合
まず $n=0$ の場合を考える．代入すると，

$$
\frac{dy}{dx} + p(x)y = q(x)
$$

となる．これは一階線形非同次微分方程式であり，積分因子法や定数変化法で解くことができる．

### $n=1$ の場合
次に $n=1$ の場合を考える．代入すると，

$$
\begin{aligned}
\frac{dy}{dx} + p(x)y &= q(x)y \\
\frac{dy}{dx} + \left(p(x) - q(x) \right)y &= 0
\end{aligned}
$$

となる．これは一階線形同次微分方程式であり，変数分離法で解くことができる．

### $n\neq0,1$ の場合
ここから $n\neq0,1$ の場合を扱う．この場合は，$y\neq 0$ の区間で

$$
z = y^{1-n}=\frac{1}{y^{n-1}}
$$

とおく．$y,z$ はどちらも $x$ の関数なので，$x$ について微分すると，

$$
\begin{aligned}
\frac{dz}{dx} &= \frac{d}{dx}y^{1-n} \\
&= \left(1-n \right)y^{-n} \frac{dy}{dx}
\end{aligned}
$$

となる．これを使って元の微分方程式を書き換える．まず両辺を $y^n$ で割ると，

$$
\begin{aligned}
\frac{dy}{dx} + p(x)y &= q(x)y^n \\
\frac{1}{y^n}\frac{dy}{dx} + p(x) \frac{1}{y^{n-1}} &= q(x)
\end{aligned}
$$

である．ここで $z=y^{1-n}$ と $\frac{dz}{dx}=(1-n)y^{-n}\frac{dy}{dx}$ を代入すると，

$$
\begin{aligned}
\frac{1}{1-n}\frac{dz}{dx} + p(x) z &= q(x) \\
\frac{dz}{dx} + (1-n)p(x)z &= (1-n)q(x)
\end{aligned}
$$

となる．これは $z$ に関する一階線形非同次微分方程式である．積分因子法を用いると，

$$
\begin{aligned}
z &= e^{-(1-n)\int p(x)\,dx}
\left\{(1-n)\int q(x)e^{(1-n)\int p(x)\,dx}\,dx + C\right\}
\end{aligned}
$$

が得られる．最後に $z=y^{1-n}$ を戻せば，$y$ は

$$
\begin{aligned}
y &= \left[
e^{-(1-n)\int p(x)\,dx}
\left\{(1-n)\int q(x)e^{(1-n)\int p(x)\,dx}\,dx + C\right\}
\right]^{\frac{1}{1-n}}
\end{aligned}
$$

と表される．

```{admonition} 変数変換で失われる可能性のある解
上の導出では $y^n$ で割り，さらに $z=y^{1-n}$ とおいたため，$y=0$ を含む解は別途確認する必要がある．特に $n>0$ のときは $y=0$ が元の方程式を満たす場合がある．
```


## ベルヌーイ型微分方程式の具体例
では，次の微分方程式を解いてみよう．

$$
\frac{dy}{dx} - y = - xy^2
$$

解答はクリックで確認できる．

````{dropdown} 解答
これは $n=2$ のベルヌーイ型であるので，$y\neq 0$ の区間で $z=y^{-1}$ として変数変換する．

$$
\begin{aligned}
\frac{dz}{dx} &= \frac{d}{dx}y^{-1} \\
&= - y^{-2} \frac{dy}{dx}
\end{aligned}
$$

元の微分方程式の両辺を $y^2$ で割り，$z$ と $\frac{dz}{dx}$ で書き換えると，

$$
\begin{aligned}
\frac{dy}{dx} - y &= -xy^2 \\
\frac{1}{y^2}\frac{dy}{dx} - \frac{1}{y} &= -x \\
-\frac{dz}{dx} - z &= -x \\
\frac{dz}{dx} + z &= x
\end{aligned}
$$

となる．これは $z$ に関する一階線形非同次微分方程式である．積分因子は $e^x$ なので，

$$
\begin{aligned}
\frac{d}{dx}\left(ze^x\right) &= xe^x \\
ze^x &= xe^x - e^x + C \\
z &= x - 1 + Ce^{-x}
\end{aligned}
$$

が得られる．$z=\frac{1}{y}$ を戻すと，

$$
\begin{aligned}
\frac{1}{y} &= x - 1 + Ce^{-x} \\
y &= \frac{1}{x - 1 + Ce^{-x}}
\end{aligned}
$$

となる．また，元の微分方程式には $y=0$ も解として含まれる．これは $y^2$ で割ったときに除外された解である．
````


## Pythonによる実装
では，上記の微分方程式を SymPy で解く．SymPy の出力は，手計算で得た非零解の分母分子に $e^x$ をかけた形になっている．

`````python
from sympy import symbols, Eq, Derivative, Function, dsolve

# 変数と未知関数の定義
x = symbols('x')
y = Function('y')(x)

# dy/dxの定義
dy = Derivative(y, x)

# 微分方程式の定義
eq = Eq(dy - y, -x*y**2)

# 微分方程式を解く
y_ = dsolve(eq, y)
y_
`````

`````python
import numpy as np
from sympy import plotting
import matplotlib.pyplot as plt

N = 10
C1 = symbols('C1')

eqs = [0]
for c in np.linspace(-5, 5, N):
    eqs.append(y_.rhs.subs(C1, c))
p = plotting.plot(*eqs, (x, 0, 5), ylim=(-5, 20), show=False)

p[0].line_color = 'black'
p[0].label = 'zero solution'
cm = plt.get_cmap('magma', N)
for i in range(1, len(eqs)):
    p[i].line_color = cm(i - 1)
p.show()
`````

黒線は，変数変換の途中で除外された零解 $y=0$ である．その他の曲線は，分母が $0$ になる点で定義されないため，グラフ上では垂直方向に発散することがある．

### NumPy + Matplotlib による可視化
SymPy の `plotting.plot` は数式を直接描画できるため便利であるが，分母が $0$ になる点の近くでは曲線が不自然につながって見えることがある．

そこで，SymPy の式を NumPy で評価できる関数に変換する `lambdify` を使い，定義されない点の近くや表示範囲を大きく外れる点を `nan` として除外して描画する．Matplotlib は `nan` を含む箇所では線をつながないため，発散点の前後を別の曲線として表示できる．

`````python
from sympy import lambdify

y_func = lambdify((x, C1), y_.rhs, 'numpy')
den_func = lambdify((x, C1), 1 / y_.rhs, 'numpy')

xs = np.linspace(0, 5, 3000)
ylim = (-5, 20)

fig, ax = plt.subplots()

# 零解
ax.plot(xs, np.zeros_like(xs), color='black')

cm = plt.get_cmap('magma', N)
for i, c in enumerate(np.linspace(-5, 5, N)):
    ys = y_func(xs, c)
    den = den_func(xs, c)
    ys = np.asarray(ys, dtype=float)

    # 発散点や表示範囲を大きく外れる点では線を切る
    ys = np.where(np.isfinite(ys), ys, np.nan)
    ys = np.where(np.abs(den) > 1e-2, ys, np.nan)
    ys = np.where((ylim[0] <= ys) & (ys <= ylim[1]), ys, np.nan)

    ax.plot(xs, ys, color=cm(i))

ax.set_xlim(0, 5)
ax.set_ylim(*ylim)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.legend()
plt.show()
`````
