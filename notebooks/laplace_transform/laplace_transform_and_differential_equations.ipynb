## ラプラス変換
$t\geq0$ で定義された関数 $f(t)$ の片側ラプラス変換は，

$$
F(s)=\mathcal L\{f(t)\}=\int_0^\infty f(t)e^{-st}dt
$$

であった．ここで $s=\sigma+i\omega$ は複素変数であり，$\operatorname{Re}(s)=\sigma$ が積分の収束を決める．像関数 $F(s)$ から原関数 $f(t)$ へ戻す操作を逆ラプラス変換といい，

$$
f(t)=\mathcal L^{-1}\{F(s)\}
$$

と表す．このノートではBromwich積分を直接計算せず，変換表と部分分数分解を用いて逆ラプラス変換を求める．

ここでは，導関数のラプラス変換によって常微分方程式の初期値問題を代数的に解く方法を紹介する．計算で用いる主な変換対と性質を以下にまとめる．

| $f(t)$ | $F(s)$ |
| :---: | :---: |
| $1$ | $\frac{1}{s}$ |
| $\delta(t)$ | $1$ |
| $e^{at}$ | $\frac{1}{s-a}$ |
| $t^{n}$ | $\frac{n!}{s^{n+1}}$ |
| $\cos(\omega t)$ | $\frac{s}{s^{2}+\omega^{2}}$ |
| $\sin(\omega t)$ | $\frac{\omega}{s^{2}+\omega^{2}}$ |
| $e^{at}\cos(\omega t)$ | $\frac{s-a}{(s-a)^{2}+\omega^{2}}$ |
| $e^{at}\sin(\omega t)$ | $\frac{\omega}{(s-a)^{2}+\omega^{2}}$ |
| $e^{at}f(t)$（$s$ 領域の移動則） | $F(s-a)$ |
| $f(at),\ a>0$（時間スケーリング） | $\frac{1}{a}F\left(\frac{s}{a}\right)$ |
| $f'(t)$ | $sF(s)-f(0+)$ |
| $f''(t)$ | $s^2F(s)-sf(0+)-f'(0+)$ |

変換が収束する範囲は関数によって異なる．例えば，$1$，$t^n$，$\sin(\omega t)$，$\cos(\omega t)$ では $\operatorname{Re}(s)>0$，$e^{at}$ では $\operatorname{Re}(s)>a$ である．

## 常微分方程式をラプラス変換で解く
常微分方程式の初期値問題をラプラス変換で解く手順は以下である．

1. 微分方程式の両辺をラプラス変換する
2. 初期値を代入し，像関数 $Y(s)$ についての代数方程式を解く
3. $Y(s)$ を部分分数分解し，逆ラプラス変換して $y(t)$ を求める

次の二階定数係数線形非同次微分方程式を考える．

$$
y''(t)-2y'(t)+y(t)=t
$$

初期値は

$$
y(0)=0, \qquad y'(0)=0
$$

とする．通常の解法では同次方程式の一般解と非同次方程式の特殊解を組み合わせてから初期値を適用するが，ラプラス変換では初期値を含めたまま代数方程式へ変換できる．

### ラプラス変換
$y(t)$ のラプラス変換を $Y(s)$ とする．線形性により，

$$
\mathcal L\{y''(t)\}-2\mathcal L\{y'(t)\}+\mathcal L\{y(t)\}=\mathcal L\{t\}
$$

と変換できる．各項は，

$$
\mathcal L\{y''(t)\}=s^2Y(s)-sy(0)-y'(0)
$$

$$
\mathcal L\{y'(t)\}=sY(s)-y(0)
$$

$$
\mathcal L\{y(t)\}=Y(s), \qquad \mathcal L\{t\}=\frac{1}{s^2}
$$

となる．したがって，微分方程式は

$$
\{s^2Y(s)-sy(0)-y'(0)\}-2\{sY(s)-y(0)\}+Y(s)=\frac{1}{s^2}
$$

へ変換される．これは，$t$ 領域の微分方程式を $s$ 領域の代数方程式へ変換したものである．

### 初期値の代入と代数処理
$y(0)=0$ と $y'(0)=0$ を代入し，$Y(s)$ について整理する．

$$
\begin{align}
s^2Y(s)-2sY(s)+Y(s)&=\frac{1}{s^2} \\
(s-1)^2Y(s)&=\frac{1}{s^2} \\
Y(s)&=\frac{1}{s^2(s-1)^2}
\end{align}
$$

### 部分分数分解
変換表を使って逆ラプラス変換できる形へ部分分数分解する．

$$
Y(s)=\frac{1}{s^2(s-1)^2}
=\frac{2}{s}+\frac{1}{s^2}-\frac{2}{s-1}+\frac{1}{(s-1)^2}
$$

### 逆ラプラス変換
線形性と変換表を用いると，

$$
\begin{align}
y(t)&=\mathcal L^{-1}\{Y(s)\} \\
&=\mathcal L^{-1}\left\{\frac{2}{s}\right\}
+\mathcal L^{-1}\left\{\frac{1}{s^2}\right\}
-\mathcal L^{-1}\left\{\frac{2}{s-1}\right\}
+\mathcal L^{-1}\left\{\frac{1}{(s-1)^2}\right\} \\
&=2+t-2e^t+te^t
\end{align}
$$

となる．以上より，初期値問題の解は

$$
y(t)=2+t-2e^t+te^t
$$

である．

```{admonition} 第4項の逆ラプラス変換
$\mathcal L\{t\}=1/s^2$ と $s$ 領域の移動則

$$
\mathcal L\{e^{at}f(t)\}=F(s-a)
$$

を用いる．$f(t)=t$，$a=1$ とすると，

$$
\mathcal L\{te^t\}=\frac{1}{(s-1)^2}
$$

であるため，$\mathcal L^{-1}\{1/(s-1)^2\}=te^t$ となる．
```

### 解の確認
得られた解について，

$$
y(0)=0, \qquad y'(0)=0
$$

が成り立つ．また，微分して元の方程式へ代入すると，

$$
y''(t)-2y'(t)+y(t)=t
$$

となるため，初期条件と微分方程式の両方を満たしている．
