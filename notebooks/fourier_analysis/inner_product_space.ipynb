ここまで離散データ，ベクトル，関数について直交性を説明してきた．直交性は「内積が $0$ である」という考え方で定義されるが，ベクトル以外の対象に対して内積を考えることに不自然さを感じた方もいるかと思う．ここでは，内積演算が満たすべき条件（**内積の公理**）を紹介し，ベクトルだけでなく離散データや関数に対しても内積を定義できることを確認する．

## 内積の公理

線形空間 $\mathcal{L}$ の元 $\mathbf{u}, \mathbf{v} \in \mathcal{L}$ に対して，2つの元から実数を返す演算 $\langle \mathbf{u}, \mathbf{v} \rangle$ が以下を満たすとき，$\langle \mathbf{u}, \mathbf{v} \rangle$ を **内積** という．ここでは実数をスカラーとする線形空間を考える．

- 正値性

    $$
    \langle \mathbf{u}, \mathbf{u} \rangle \geq 0
    $$

    さらに，$\langle \mathbf{u}, \mathbf{u} \rangle = 0$ となるのは $\mathbf{u}=\mathbf{0}$ のときに限る．

- 対称性

    $$
    \langle \mathbf{u}, \mathbf{v} \rangle = \langle \mathbf{v}, \mathbf{u} \rangle
    $$

- 線形性

    任意のスカラー $c_1, c_2 \in \mathbb{R}$ と任意の元 $\mathbf{u}_1, \mathbf{u}_2, \mathbf{v} \in \mathcal{L}$ について，

    $$
    \langle c_1 \mathbf{u}_1 + c_2 \mathbf{u}_2, \mathbf{v} \rangle = c_1 \langle \mathbf{u}_1, \mathbf{v} \rangle + c_2 \langle \mathbf{u}_2, \mathbf{v} \rangle
    $$

    が成り立つ．

上記の条件を満たす線形空間を **内積空間** という．内積空間では，内積を用いてノルムを定義できる．

$$
\left \| \mathbf{u} \right \| = \sqrt{\langle \mathbf{u}, \mathbf{u} \rangle}
$$

さらに，ノルムを用いると2つの元の距離を

$$
d(\mathbf{u}, \mathbf{v}) = \left \| \mathbf{u} - \mathbf{v} \right \|
$$

と定義できる．そのため，内積空間は距離をもつ空間（計量空間）としても見ることができる．

内積空間の任意の元 $\mathbf{u}, \mathbf{v} \in \mathcal{L}$ について，

$$
\langle \mathbf{u}, \mathbf{v} \rangle = 0
$$

が成り立つとき，これらの元は **直交する** という．

## 内積の例

内積はベクトルだけでなく，離散データや関数に対しても定義できる．どの場合も「対応する成分どうしを掛け合わせて足し合わせる」という考え方が基本になっている．

### ベクトルの内積

$m$ 次元ベクトル $\mathbf{u}, \mathbf{v} \in \mathbb{R}^{m}$ に対して，通常の内積は以下で定義される．

$$
\langle \mathbf{u}, \mathbf{v} \rangle = \sum_{i=1}^{m} u_i v_i
$$

この内積を用いると，ベクトルのノルムは

$$
\left \| \mathbf{u} \right \| = \sqrt{\sum_{i=1}^{m} u_i^2}
$$

となる．

### 離散データの内積

$N$ 個の点 $x_1, x_2, \ldots, x_N$ で観測された2つのデータ列 $f(x_\alpha), g(x_\alpha)$ を考える．このとき，離散データの内積を

$$
\langle f, g \rangle_N = \sum_{\alpha=1}^{N} f(x_\alpha)g(x_\alpha)
$$

と定義できる．これは，前節で正規方程式を考えたときに現れた

$$
\sum_{\alpha=1}^{N} \phi_i(x_\alpha)\phi_j(x_\alpha)
$$

という量そのものである．したがって，

$$
\langle \phi_i, \phi_j \rangle_N = 0 \qquad (i \neq j)
$$

が成り立つとき，離散データとしての関数 $\phi_i, \phi_j$ は直交していると考えられる．

### 関数の内積

区間 $[a,b]$ 上の関数 $f(x), g(x)$ に対して，関数の内積を以下で定義できる．

$$
\langle f, g \rangle = \int_a^b f(x)g(x) dx
$$

この内積が定義できるためには，少なくとも $\int_a^b f(x)^2 dx$ や $\int_a^b g(x)^2 dx$ が有限である必要がある．本講義で扱う連続関数では，この条件は通常満たされる．

この関数の内積を用いると，関数のノルムは

$$
\left \| f \right \| = \sqrt{\int_a^b f(x)^2 dx}
$$

となる．また，

$$
\langle f, g \rangle = \int_a^b f(x)g(x) dx = 0
$$

が成り立つとき，区間 $[a,b]$ 上で $f(x)$ と $g(x)$ は直交するという．

## 直交系での係数

関数 $f(x)$ を，互いに直交する関数系 $\{\phi_i(x)\}$ の線形結合で近似することを考える．

$$
f(x) \approx \sum_i c_i \phi_i(x)
$$

このとき，直交性を用いると係数 $c_i$ は

$$
c_i = \frac{\langle f, \phi_i \rangle}{\langle \phi_i, \phi_i \rangle} = \frac{\langle f, \phi_i \rangle}{\left \| \phi_i \right \|^2}
$$

と表せる．この式は，次に学ぶフーリエ級数展開のフーリエ係数を導出するときの基本となる．

```{admonition} 複素数値関数の内積
複素フーリエ級数では，$e^{ikx}$ のような複素数値関数を扱う．複素数値関数の場合，内積は通常，片方に複素共役をつけて

$$
\langle f, g \rangle = \int_a^b f(x)\overline{g(x)} dx
$$

のように定義する．このとき対称性は $\langle f, g \rangle = \overline{\langle g, f \rangle}$ という形になる．
```
